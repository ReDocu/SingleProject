```python
# Crawling
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

start_level = 0
end_level = 30

problem_dict = {}

columns = ['p_id','p_name','p_href','p_level','p_answer_people','try_count','category']

problem_df = pd.DataFrame(columns=columns)

for level in range(start_level,end_level + 1):
    
    url = 'https://solved.ac/problems/level/{}?page=1'.format(level)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    maxpage = int(soup.select('div.css-18lc7iz > a')[-1].text.strip())
    
    for page in range(1, maxpage + 1):
        print(page)
        url = 'https://solved.ac/problems/level/{}?page={}'.format(level, page)

        res = requests.get(url)
        soup = BeautifulSoup(res.text, 'html.parser')

        path = 'tbody > tr'
        
        for item in soup.select(path):
            number = 0
            name = ''
            answer_people = 0
            try_count = 0.0
            href = ''
            # a 태그 찾기
            try:
                number = item.select('td')[0].select('a')[0].text
            except:
                print('0번 a태그가 존재하지 않습니다.')

            try:
                href = item.select('td')[0].select('a')[0]['href']
            except:
                print('하이퍼링크를 찾지 못했습니다.')    

            # a 태그 찾기
            try:
                name = item.select('td')[1].select('a')[0].text
            except:
                print('1번 a태그가 존재하지 않습니다.')

            try:
                answer_people = item.select('td')[2].select('div')[0].text
            except:
                print('2번 div 태그가 존재하지 않습니다.')

            try:
                try_count = item.select('td')[3].select('div')[0].text
            except:
                print('3번 div 태그가 존재하지 않습니다.')
                
                
            problem_df.loc[len(problem_df)] = [number,name,href,level,answer_people,try_count,np.NaN]

        
problem_df
```

```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd

class PandasGUI:
    def __init__(self, master):
        self.master = master
        self.master.title("Pandas GUI")

        # 파일 선택 버튼
        self.file_button = tk.Button(self.master, text="파일 선택", command=self.load_file)
        self.file_button.pack(pady=10)

        # 통계 정보 표시 레이블
        self.stats_label = tk.Label(self.master, text="")
        self.stats_label.pack(pady=10)

    def load_file(self):
        file_path = filedialog.askopenfilename(filetypes=[("CSV files", "*.csv")])

        if file_path:
            # CSV 파일을 Pandas DataFrame으로 로드
            try:
                df = pd.read_csv(file_path)
                # 간단한 통계 정보 표시
                stats = df.describe()
                self.stats_label.config(text=stats)
            except Exception as e:
                self.stats_label.config(text=f"오류: {str(e)}")

if __name__ == "__main__":
    root = tk.Tk()
    app = PandasGUI(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd

class PandasGUI:
    def __init__(self, master):
        self.master = master
        self.master.title("Pandas GUI")

        # 파일 선택 버튼
        self.file_button = tk.Button(self.master, text="파일 선택", command=self.load_file)
        self.file_button.pack(pady=10)

        # 편집 텍스트 박스
        self.edit_text = tk.Text(self.master, height=10, width=40)
        self.edit_text.pack(pady=10)

        # 저장 버튼
        self.save_button = tk.Button(self.master, text="다른 이름으로 저장", command=self.save_as)
        self.save_button.pack(pady=10)

        # 현재 데이터프레임
        self.df = pd.DataFrame()

    def load_file(self):
        file_path = filedialog.askopenfilename(filetypes=[("CSV files", "*.csv")])

        if file_path:
            try:
                # CSV 파일을 Pandas DataFrame으로 로드
                self.df = pd.read_csv(file_path)
                # 데이터프레임을 텍스트 박스에 표시
                self.edit_text.delete(1.0, tk.END)
                self.edit_text.insert(tk.END, str(self.df))
            except Exception as e:
                self.edit_text.delete(1.0, tk.END)
                self.edit_text.insert(tk.END, f"오류: {str(e)}")

    def save_as(self):
        file_path = filedialog.asksaveasfilename(defaultextension=".csv", filetypes=[("CSV files", "*.csv")])

        if file_path:
            try:
                # 현재 데이터프레임을 선택한 경로에 저장
                self.df.to_csv(file_path, index=False)
                tk.messagebox.showinfo("저장 완료", "파일이 성공적으로 저장되었습니다.")
            except Exception as e:
                tk.messagebox.showerror("오류", f"저장 중 오류 발생: {str(e)}")

if __name__ == "__main__":
    root = tk.Tk()
    app = PandasGUI(root)
    root.mainloop()
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

url = 'https://musescore.com/sheetmusic/pop?page=1'

driver = webdriver.Chrome()
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

music_list = []
music_fail_list = []
for music_down in range(1,101):
    temp_url = f'https://musescore.com/sheetmusic/pop?page={music_down}' 
    driver.get(temp_url)
    time.sleep(1)
    
    for item in soup.select('body > div.js-page.react-container > div.b1GqE > section > section > main > div > section > article > div > div.EzJvq > a'):
        
        try:
            music_list.append(item['href'])
        except:
            music_fail_list.append(item)
    print(music_down, ":" , len(music_list))
    
f = open('music_list.txt','w')

for item in music_list:
    f.write(item)
    f.write("\n")

f.close()
```

# 게임정보 가져오는 코드

```python
import os
import re

import warnings
warnings.filterwarnings(action='ignore')

import pandas as pd

import DFMaker

# 데이터 프레임 구성하기
game_info_columns = [
    'AppID',
    'Name',
    'Developer',
    'Company',
    'price',
    'latest rating',
    'latest rating counter',
    'all rating',
    'all rating counter',
    'release date',
    'category',
    'DLC_Game']


csv_root_path = 'game_data'
csv_folder_list = os.listdir(csv_root_path)

# csv 파일 불러와서 체크하기
try:
    game_df = pd.read_csv(r'static_data\game_info.csv')
except:
    game_df = pd.DataFrame(columns=game_info_columns)

try:
    game_df = game_df.drop(columns=['Unnamed: 0.1','Unnamed: 0'])
except:
    pass

file_list = []

# 존재하는지 비교하기
for item in csv_folder_list:
    reg = re.compile('\d+')
    file = reg.findall(item)

    if (game_df['AppID'] == int(file[0])).sum() > 0:
        continue

    file_list.append(file[0])



new_df = pd.DataFrame(columns=game_info_columns)

error_list = []
# 게임 추가해서 더하기
for item in file_list:
    try:
        game_info_df = DFMaker.create_game_info(item)
        new_df = pd.concat([new_df, game_info_df])
    except:
        print('{} error'.format(item))
        error_list.append(item)

game_df = pd.concat([game_df,new_df])
game_df.to_csv('static_data\game_info.csv')

```

# 게임 정보 + 가격 정보 합본

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from bs4 import BeautifulSoup

import pandas as pd

driver = webdriver.Chrome()

account_columns = ['AppID','Total_Gross', 'Total_Net','Total_owners']

# 게임 정보 가져오기
game_info_df = pd.read_csv('static_data\game_info.csv')
game_info_df['AppID'].sort_values()

# Account 정보 가져오기
try:
    account_df = pd.read_csv('static_data\game_account_info.csv')
except:
    account_df = pd.DataFrame(columns=account_columns)

# 작업 완료된 데이터 필터링
add_file_list = []
error_file_list = []

for game_id in game_info_df['AppID'].sort_values().values:
    if (account_df['AppID'] == game_id).sum() > 0:
        continue

    print('작업 게임 ID : ', game_id)

    try:
        url = 'https://gamesensor.info/app/{}'.format(game_id)
        driver.get(url)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # 데이터 파싱
        total_gross = ''
        total_net = ''
        total_owners = ''
        for item in soup.select('#user_page > div.col-lg-6'):
            if str(item.select('div.text')[0]).find('Total Gross') != -1:
                recv = item.select('div > div.value')[0].text.strip()
                total_gross = recv.split('$')[1]
                total_net = recv.split('$')[2]
            if str(item.select('div.text')[0]).find('Total Owners') != -1:
                total_owners = item.select('div > div.value')[0].text.strip()

        result_total_gross = 0
        result_total_net = 0
        result_total_owners = 0

        total_gross = total_gross.replace('>','')
        total_gross = total_gross.replace('$','')

        total_net = total_net.replace('>','')
        total_net = total_net.replace('$','')

        total_owners = total_owners.replace('>','')
        total_owners = total_owners.replace('$','')

        if 'M' in total_gross:
            result_total_gross = float(total_gross.replace('M','')) * 1000000
        if 'K' in total_gross:
            result_total_gross =float(total_gross.replace('K','')) * 1000
        if 'M' in total_net:
            result_total_net = float(total_net.replace('M','')) * 1000000
        if 'K' in total_net:
            result_total_net =float(total_net.replace('K','')) * 1000
        if 'M' in total_owners:
            result_total_owners = float(total_owners.replace('M','')) * 1000000
        if 'K' in total_owners:
            result_total_owners =float(total_owners.replace('K','')) * 1000

        # 데이터 추가하기
        new_account_df = pd.DataFrame(columns=account_columns)
        new_account_df.loc[0] = [game_id, result_total_gross, result_total_net, result_total_owners]
        account_df = pd.concat([account_df, new_account_df])
        account_df.to_csv('static_data\game_account_info.csv')
    except:
        # error 파일 검증
        error_file_list.append(game_id)
        print('Error : ',game_id)

# 저장
account_df.to_csv('static_data\game_account_info.csv')

print(error_file_list)
```

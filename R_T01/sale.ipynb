```python
res = '''
<body class="page-sales_history" itemscope="" itemtype="http://schema.org/WebSite" data-steam-cdn="https://cdn.cloudflare.steamstatic.com/" data-rev="acfac9ba89">
<a class="skip-main" href="#main" id="js-skip-main">Skip to content</a>
<div class="header" role="banner">
<div class="header-menu-button-container">
<button type="button" class="header-menu-button" id="js-header-hamburger" accesskey="m" aria-label="Toggle menu"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-three-bars" aria-hidden="true"><path d="M1 2.75A.75.75 0 0 1 1.75 2h12.5a.75.75 0 0 1 0 1.5H1.75A.75.75 0 0 1 1 2.75Zm0 5A.75.75 0 0 1 1.75 7h12.5a.75.75 0 0 1 0 1.5H1.75A.75.75 0 0 1 1 7.75ZM1.75 12h12.5a.75.75 0 0 1 0 1.5H1.75a.75.75 0 0 1 0-1.5Z"></path></svg> <span>Menu</span></button>
</div>
<div class="header-navbar">
<a href="/" class="header-brand" itemprop="url"><svg width="30" height="30" viewBox="0 0 128 128" class="octicon octicon-steamdb" aria-hidden="true"><path fill-rule="evenodd" d="M63.9 0C30.5 0 3.1 11.9.1 27.1l35.6 6.7c2.9-.9 6.2-1.3 9.6-1.3l16.7-10c-.2-2.5 1.3-5.1 4.7-7.2 4.8-3.1 12.3-4.8 19.9-4.8 5.2-.1 10.5.7 15 2.2 11.2 3.8 13.7 11.1 5.7 16.3-5.1 3.3-13.3 5-21.4 4.8l-22 7.9c-.2 1.6-1.3 3.1-3.4 4.5-5.9 3.8-17.4 4.7-25.6 1.9-3.6-1.2-6-3-7-4.8L2.5 38.4c2.3 3.6 6 6.9 10.8 9.8C5 53 0 59 0 65.5c0 6.4 4.8 12.3 12.9 17.1C4.8 87.3 0 93.2 0 99.6 0 115.3 28.6 128 64 128c35.3 0 64-12.7 64-28.4 0-6.4-4.8-12.3-12.9-17 8.1-4.8 12.9-10.7 12.9-17.1 0-6.5-5-12.6-13.4-17.4 8.3-5.1 13.3-11.4 13.3-18.2 0-16.5-28.7-29.9-64-29.9zm22.8 14.2c-5.2.1-10.2 1.2-13.4 3.3-5.5 3.6-3.8 8.5 3.8 11.1 7.6 2.6 18.1 1.8 23.6-1.8s3.8-8.5-3.8-11c-3.1-1-6.7-1.5-10.2-1.5zm.3 1.7c7.4 0 13.3 2.8 13.3 6.2 0 3.4-5.9 6.2-13.3 6.2s-13.3-2.8-13.3-6.2c0-3.4 5.9-6.2 13.3-6.2zM45.3 34.4c-1.6.1-3.1.2-4.6.4l9.1 1.7a10.8 5 0 1 1-8.1 9.3l-8.9-1.7c1 .9 2.4 1.7 4.3 2.4 6.4 2.2 15.4 1.5 20-1.5s3.2-7.2-3.2-9.3c-2.6-.9-5.7-1.3-8.6-1.3zM109 51v9.3c0 11-20.2 19.9-45 19.9-24.9 0-45-8.9-45-19.9v-9.2c11.5 5.3 27.4 8.6 44.9 8.6 17.6 0 33.6-3.3 45.2-8.7zm0 34.6v8.8c0 11-20.2 19.9-45 19.9-24.9 0-45-8.9-45-19.9v-8.8c11.6 5.1 27.4 8.2 45 8.2s33.5-3.1 45-8.2z"></path></svg> <span>SteamDB</span></a>
<form class="header-search" id="js-header-form" method="get" action="/search/" itemprop="potentialAction" itemscope="" itemtype="http://schema.org/SearchAction" role="search">
<input type="hidden" name="a" value="all">
<meta itemprop="target" content="https://steamdb.info/search/?a=all&amp;q={q}">
<input type="text" accesskey="/" class="field" autocomplete="off" name="q" placeholder="Searchâ€¦" itemprop="query-input" aria-label="Search">
<button type="submit" class="submit" aria-label="Perform search" tabindex="-1"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-search" aria-hidden="true"><path d="M10.68 11.74a6 6 0 0 1-7.922-8.982 6 6 0 0 1 8.982 7.922l3.04 3.04a.749.749 0 0 1-.326 1.275.749.749 0 0 1-.734-.215ZM11.5 7a4.499 4.499 0 1 0-8.997 0A4.499 4.499 0 0 0 11.5 7Z"></path></svg></button>
</form>
<a href="/sales/">Sales</a>
<a href="/charts/">Charts</a>
<a href="/calculator/">Calculator</a>
<a href="/calendar/">Calendar</a>
<a href="/patchnotes/">Patches</a>
<a href="/discord/">Discord</a>
<a href="/instantsearch/">Find games</a>
<button type="button" class="btn btn-outline header-search-close">Close</button>
</div>
<button type="button" class="header-search-mobile" aria-label="Open search"><svg width="24" height="24" viewBox="0 0 16 16" class="octicon octicon-search" aria-hidden="true"><path d="M10.68 11.74a6 6 0 0 1-7.922-8.982 6 6 0 0 1 8.982 7.922l3.04 3.04a.749.749 0 0 1-.326 1.275.749.749 0 0 1-.734-.215ZM11.5 7a4.499 4.499 0 1 0-8.997 0A4.499 4.499 0 0 0 11.5 7Z"></path></svg></button>
<div class="header-user hide-small">
<a class="header-user-avatar" href="/calculator/76561199195333179/" aria-label="User menu">
<svg width="20" height="20" viewBox="0 0 16 16" class="octicon octicon-triangle-down" aria-hidden="true"><path d="m4.427 7.427 3.396 3.396a.25.25 0 0 0 .354 0l3.396-3.396A.25.25 0 0 0 11.396 7H4.604a.25.25 0 0 0-.177.427Z"></path></svg><img class="avatar" src="/static/img/appicon.svg" alt="" width="30" height="30">
</a>
</div>
</div>
<div class="header-search-suggestions">
<div class="container" id="js-search-tips">
<p>Start typing to see game suggestions. This only suggests apps that have a store page.</p>
<ul>
<li class="hide-small">Use slash key (<kbd>/</kbd>) to focus search from anywhere.</li>
<li class="hide-small">Use arrow keys (<kbd>â†‘</kbd> and <kbd>â†“</kbd>) to navigate suggestions.</li>
<li class="hide-small">Use escape (<kbd>Esc</kbd>) to close search.</li>
<li>Enter an appid to be redirected to the app page.</li>
<li>Enter a steamid (765...) to be redirected to calculator.</li>
<li>Paste a profile link (/id/ or /profiles/) to be redirected to calculator.</li>
<li>Paste any url that contains app/, sub/, bundle/, or depot/ to be redirected to the relevant page.</li>
<li><a href="https://www.algolia.com/?utm_content=steamdb.info&amp;utm_campaign=poweredby">Search is powered by Algolia.</a></li>
</ul>
</div>
<div class="container" id="js-search-suggestions"></div>
<div class="container header-search-enter">
Only apps with a store page are suggested.
<span class="hide-small">
<kbd>Enter</kbd> Submit to view all results.
<kbd>Ctrl+Enter</kbd> View and filter in instant search.
</span>
</div>
<div class="header-search-side-close">
<div class="header-search-side-close-inner">Close Ã—</div>
</div>
</div>
<div class="header-menu-container" hidden="">
<div class="header-menu" tabindex="0">
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-info" aria-hidden="true"><path d="M0 8a8 8 0 1 1 16 0A8 8 0 0 1 0 8Zm8-6.5a6.5 6.5 0 1 0 0 13 6.5 6.5 0 0 0 0-13ZM6.5 7.75A.75.75 0 0 1 7.25 7h1a.75.75 0 0 1 .75.75v2.75h.25a.75.75 0 0 1 0 1.5h-2a.75.75 0 0 1 0-1.5h.25v-2h-.25a.75.75 0 0 1-.75-.75ZM8 6a1 1 0 1 1 0-2 1 1 0 0 1 0 2Z"></path></svg> About</li>
<li><a href="/blog/">Blog</a></li>
<li><a href="/discord/">Discord</a></li>
<li><a href="/discord/bot/">Discord Bot</a></li>
<li><a href="/faq/">FAQ &amp; help</a></li>
<li><a href="/donate/">Donate &amp; contribute</a></li>
<li><a href="/extension/">Browser extension</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-search" aria-hidden="true"><path d="M10.68 11.74a6 6 0 0 1-7.922-8.982 6 6 0 0 1 8.982 7.922l3.04 3.04a.749.749 0 0 1-.326 1.275.749.749 0 0 1-.734-.215ZM11.5 7a4.499 4.499 0 1 0-8.997 0A4.499 4.499 0 0 0 11.5 7Z"></path></svg> Tools</li>
<li><a href="/sales/">Sales</a></li>
<li><a href="/charts/">Charts</a></li>
<li><a href="/calculator/">Calculator</a></li>
<li><a href="/patchnotes/">Patch notes</a></li>
<li><a href="/search/">Search</a></li>
<li><a href="/instantsearch/">Instant search</a></li>
<li><a href="/sales/history/">When is next sale?</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-sync" aria-hidden="true"><path d="M1.705 8.005a.75.75 0 0 1 .834.656 5.5 5.5 0 0 0 9.592 2.97l-1.204-1.204a.25.25 0 0 1 .177-.427h3.646a.25.25 0 0 1 .25.25v3.646a.25.25 0 0 1-.427.177l-1.38-1.38A7.002 7.002 0 0 1 1.05 8.84a.75.75 0 0 1 .656-.834ZM8 2.5a5.487 5.487 0 0 0-4.131 1.869l1.204 1.204A.25.25 0 0 1 4.896 6H1.25A.25.25 0 0 1 1 5.75V2.104a.25.25 0 0 1 .427-.177l1.38 1.38A7.002 7.002 0 0 1 14.95 7.16a.75.75 0 0 1-1.49.178A5.5 5.5 0 0 0 8 2.5Z"></path></svg> Tracking</li>
<li><a href="/apps/">Apps</a></li>
<li><a href="/subs/">Packages</a></li>
<li><a href="/bundles/">Bundles</a></li>
<li><a href="/history/">History</a></li>
<li><a href="/pricechanges/">Price changes</a></li>
<li><a href="/upcoming/free/">Free promotions</a></li>
<li><a href="/tech/">Technologies &amp; built with</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-database" aria-hidden="true"><path d="M1 3.5c0-.626.292-1.165.7-1.59.406-.422.956-.767 1.579-1.041C4.525.32 6.195 0 8 0c1.805 0 3.475.32 4.722.869.622.274 1.172.62 1.578 1.04.408.426.7.965.7 1.591v9c0 .626-.292 1.165-.7 1.59-.406.422-.956.767-1.579 1.041C11.476 15.68 9.806 16 8 16c-1.805 0-3.475-.32-4.721-.869-.623-.274-1.173-.62-1.579-1.04-.408-.426-.7-.965-.7-1.591Zm1.5 0c0 .133.058.318.282.551.227.237.591.483 1.101.707C4.898 5.205 6.353 5.5 8 5.5c1.646 0 3.101-.295 4.118-.742.508-.224.873-.471 1.1-.708.224-.232.282-.417.282-.55 0-.133-.058-.318-.282-.551-.227-.237-.591-.483-1.101-.707C11.102 1.795 9.647 1.5 8 1.5c-1.646 0-3.101.295-4.118.742-.508.224-.873.471-1.1.708-.224.232-.282.417-.282.55Zm0 4.5c0 .133.058.318.282.551.227.237.591.483 1.101.707C4.898 9.705 6.353 10 8 10c1.646 0 3.101-.295 4.118-.742.508-.224.873-.471 1.1-.708.224-.232.282-.417.282-.55V5.724c-.241.15-.503.286-.778.407C11.475 6.68 9.805 7 8 7c-1.805 0-3.475-.32-4.721-.869a6.15 6.15 0 0 1-.779-.407Zm0 2.225V12.5c0 .133.058.318.282.55.227.237.592.484 1.1.708 1.016.447 2.471.742 4.118.742 1.647 0 3.102-.295 4.117-.742.51-.224.874-.47 1.101-.707.224-.233.282-.418.282-.551v-2.275c-.241.15-.503.285-.778.406-1.247.549-2.917.869-4.722.869-1.805 0-3.475-.32-4.721-.869a6.327 6.327 0 0 1-.779-.406Z"></path></svg> Database</li>
<li><a href="/tags/">User tags</a></li>
<li><a href="/calendar/">Release calendar</a></li>
<li><a href="/upcoming/">Upcoming releases</a></li>
<li><a href="/upcoming/mostfollowed/">Most followed upcoming</a></li>
<li><a href="/stats/releases/">Game releases by year</a></li>
<li><a href="/badges/">Profile badges</a></li>
<li><a href="/developers/">Developers</a></li>
<li><a href="/publishers/">Publishers</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-number" aria-hidden="true"><path d="M9 4.75A.75.75 0 0 1 9.75 4h4a.75.75 0 0 1 .53 1.28l-1.89 1.892c.312.076.604.18.867.319.742.391 1.244 1.063 1.244 2.005 0 .653-.231 1.208-.629 1.627-.386.408-.894.653-1.408.777-1.01.243-2.225.063-3.124-.527a.751.751 0 0 1 .822-1.254c.534.35 1.32.474 1.951.322.306-.073.53-.201.67-.349.129-.136.218-.32.218-.596 0-.308-.123-.509-.444-.678-.373-.197-.98-.318-1.806-.318a.75.75 0 0 1-.53-1.28l1.72-1.72H9.75A.75.75 0 0 1 9 4.75Zm-3.587 5.763c-.35-.05-.77.113-.983.572a.75.75 0 1 1-1.36-.632c.508-1.094 1.589-1.565 2.558-1.425 1 .145 1.872.945 1.872 2.222 0 1.433-1.088 2.192-1.79 2.681-.308.216-.571.397-.772.573H7a.75.75 0 0 1 0 1.5H3.75a.75.75 0 0 1-.75-.75c0-.69.3-1.211.67-1.61.348-.372.8-.676 1.15-.92.8-.56 1.18-.904 1.18-1.474 0-.473-.267-.69-.587-.737ZM5.604.089A.75.75 0 0 1 6 .75v4.77h.711a.75.75 0 0 1 0 1.5H3.759a.75.75 0 0 1 0-1.5H4.5V2.15l-.334.223a.75.75 0 0 1-.832-1.248l1.5-1a.75.75 0 0 1 .77-.037Z"></path></svg> Rankings</li>
<li><a href="/stats/gameratings/">Top rated games</a></li>
<li><a href="/badge/13/">Top game owners</a></li>
<li><a href="/stats/toplevels/">Top profile levels</a></li>
<li><a href="/stats/globaltopsellers/">Top selling games</a></li>
<li><a href="/topsellers/">Top selling games last week</a></li>
<li><a href="/stats/mostfollowed/">Most followed games</a></li>
<li><a href="/stats/mostwished/">Most wishlisted games</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-person" aria-hidden="true"><path d="M10.561 8.073a6.005 6.005 0 0 1 3.432 5.142.75.75 0 1 1-1.498.07 4.5 4.5 0 0 0-8.99 0 .75.75 0 0 1-1.498-.07 6.004 6.004 0 0 1 3.431-5.142 3.999 3.999 0 1 1 5.123 0ZM10.5 5a2.5 2.5 0 1 0-5 0 2.5 2.5 0 0 0 5 0Z"></path></svg> Account</li>
<li><a href="/watching/1235067451/">Watch list</a></li>
<li><a href="/sales/?displayOnly=OwnedGames&amp;min_reviews=0&amp;min_rating=0&amp;min_discount=0">Your games</a></li>
<li><a href="/freepackages/">Free packages</a></li>
<li><a href="/user/">Settings</a></li>
<li><a href="/signout/">Sign out</a></li>
<li>&nbsp;</li>
<li><a href="https://steamstat.us/">Steam Status</a></li>
<li><a href="https://steamapi.xpaw.me/">Steam Web API</a></li>
</ul>
<div class="header-disclaimer">SteamDB is a hobby project and is not affiliated with Valve or Steam.<br>All times on the site are UTC.</div>
<div class="header-kbd hide-small"><kbd>M</kbd> toggle this menu <kbd>/</kbd> focus search</div>
</div>
</div>
<div class="header-menu account-menu" hidden="">
<ul>
<li><a href="/calculator/76561199195333179/">Your calculator</a></li>
<li><a href="/sales/?displayOnly=OwnedGames&amp;min_reviews=0&amp;min_rating=0&amp;min_discount=0">Your games</a></li>
<li><a href="/watching/1235067451/">Watch list</a></li>
<li><a href="/user/">Settings</a></li>
<li><a href="/signout/">Sign out</a></li>
</ul>
</div>
<script nonce="" integrity="sha512-WGfIeUTU1b+rldZDI5MruL8/Axfq70QEzsMYPdQMqB2/GZ2tEt/HU2lMXmfeWiiLpaktoUlYi19Hx+Kq9YV6JA==" src="/static/js/vendor/bugsnag.js?v=5867c87944d4d5bf"></script>
<script nonce="" integrity="sha512-KdH8H31M4uxaqRFt46s+kEbCNuL1B1DeIE3UB8aUwAuPRK+fjruAaqsMGBanMWTp4nG3L9jlH1JU8/GHbA32QA==" src="/static/js/hover.js?v=29d1fc1f7d4ce2ec" defer=""></script>
<script nonce="" integrity="sha512-1ybGGSXu45WDVLpQVF82WqtnS2bxLx/FPHJTns9Vs3s6fB+Z9IRphuQHYGo6iv2bFbVXVRuQ2xNJLdKPOMw+bw==" src="/static/js/toast.js?v=d726c61925eee395" defer=""></script>
<script nonce="" integrity="sha512-54Asi0fh+BQyjP2G5y135RRYrHtTIQJ8JG8I2hH/RB/PR76OwQKfCMave0EwftDgsk8ol/qmj/Q2phGg8kxf9A==" src="/static/js/footer.js?v=e7802c8b47e1f814" defer=""></script>
<script nonce="" integrity="sha512-QOQ9LvP4di5kmch82SfxzZfqPTyTctVkDJlKU7Jato2NDbe/PydCZkoUt4nsZmvE7ZVipVMibnUoaNq/YGLREQ==" src="/static/js/global.js?v=40e43d2ef3f8762e"></script>
<div class="body-wrap">
<div class="body-content" id="main">
<script nonce="">
(function()
{
let now = Date.now();
if( 'performance' in window && 'timing' in window.performance )
{
now -= now - window.performance.timing.responseStart;
}
const timeDiff = Math.abs( now / 1000 - 1706575523 );
if( timeDiff > 10.0 )
{
window.addEventListener( 'DOMContentLoaded', () =>
{
const alert = window.SteamDB.Alert(
`Your computer time does not match our server time, countdown will be wrong. (${timeDiff.toFixed(2)}s)`,
'Is your clock out of sync?'
);
const btn = document.createElement( 'a' );
btn.className = 'btn btn-sm btn-primary';
btn.href = 'https://time.is/';
btn.textContent = 'Check your clock on https://time.is';
alert.append( btn );
} );
}
}());
</script>
<div class="next-sale-and-fest">
<div class="next-sale">
<div class="sale-discounts">
<div class="sale-discount">-30%</div>
<div class="sale-discount">-20%</div>
<div class="sale-discount">-69%</div>
<div class="sale-discount">-10%</div>
<div class="sale-discount">-95%</div>
<div class="sale-discount">-15%</div>
<div class="sale-discount">-70%</div>
<div class="sale-discount">-10%</div>
<div class="sale-discount">-10%</div>
<div class="sale-discount">-85%</div>
<div class="sale-discount">-80%</div>
<div class="sale-discount">-45%</div>
</div>
<div class="container">
<div class="sale-countdown">
<h2>When is the next major Steam sale?</h2>
<h3>
<span class="sale-name">Spring Sale 2024</span> on 14 March 2024
</h3>
<div class="sale-local-time" id="js-sale-local-time">Your local time: 15 March 2024 at 02:00</div>
<span class="huge-countdown" id="js-sale-countdown" data-target="1710435600000">44 : 16 : 14 : 05</span>
</div>
<div class="sale-countdown-now" hidden="">
<img class="huge-countdown-img" src="https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg" alt="">
<h3><span class="sale-name">Spring Sale 2024</span> is currently live and runs until 21 March 2024</h3>
<div class="small-countdown" id="js-sale-countdown-end" data-target="1711040400000"></div>
<a href="/sales/" class="huge-countdown">Browse sales Â»</a>
<div class="badge-sale-started">
ðŸ¤— The sale just started, please be patient while all the deals are updated. If some games are missing, check back soon.
</div>
</div>
</div>
</div>
<script nonce="" integrity="sha512-ZxHIJxsu3mqCW3hPp7lorM5SsPK+vIDTyruxFmdwdJpkGzZbWQcFBN/qqGo0oI5VAAzeromJ9VZOE1/5GslPtA==" src="/static/js/sales_countdown.js?v=6711c8271b2ede6a"></script>
<div class="next-fest">
<div class="container">
<div class="next-fest-title">
Next Steam fest is
<span class="sale-name">Next Fest: February 2024</span> on 5 February 2024
<time class="future-sale-time" datetime="2024-02-05T18:00:00+00:00" title="5 February 2024 at 18:00:00 UTC">in 7 days</time>
</div>
<div class="badge-sales-hints">
<a href="/sales/" class="badge-sales-hint"><svg width="24" height="24" viewBox="0 0 16 16" class="octicon octicon-project-symlink" aria-hidden="true"><path d="M0 1.75C0 .784.784 0 1.75 0h12.5C15.216 0 16 .784 16 1.75v12.5A1.75 1.75 0 0 1 14.25 16h-8.5a.75.75 0 0 1 0-1.5h8.5a.25.25 0 0 0 .25-.25V6.5h-13v1.75a.75.75 0 0 1-1.5 0ZM6.5 5h8V1.75a.25.25 0 0 0-.25-.25H6.5Zm-5 0H5V1.5H1.75a.25.25 0 0 0-.25.25Z"></path><path d="M1.5 13.737a2.25 2.25 0 0 1 2.262-2.25L4 11.49v1.938c0 .218.26.331.42.183l2.883-2.677a.25.25 0 0 0 0-.366L4.42 7.89a.25.25 0 0 0-.42.183V9.99l-.23-.001A3.75 3.75 0 0 0 0 13.738v1.012a.75.75 0 0 0 1.5 0v-1.013Z"></path></svg> <span>Browse all <u>discounted games</u> using our sales page</span></a>
<a href="/extension/" class="badge-sales-hint"><svg width="24" height="24" viewBox="0 0 16 16" class="octicon octicon-desktop-download" aria-hidden="true"><path d="m4.927 5.427 2.896 2.896a.25.25 0 0 0 .354 0l2.896-2.896A.25.25 0 0 0 10.896 5H8.75V.75a.75.75 0 1 0-1.5 0V5H5.104a.25.25 0 0 0-.177.427Z"></path><path d="M1.573 2.573a.25.25 0 0 0-.073.177v7.5a.25.25 0 0 0 .25.25h12.5a.25.25 0 0 0 .25-.25v-7.5a.25.25 0 0 0-.25-.25h-3a.75.75 0 1 1 0-1.5h3A1.75 1.75 0 0 1 16 2.75v7.5A1.75 1.75 0 0 1 14.25 12h-3.727c.099 1.041.52 1.872 1.292 2.757A.75.75 0 0 1 11.25 16h-6.5a.75.75 0 0 1-.565-1.243c.772-.885 1.192-1.716 1.292-2.757H1.75A1.75 1.75 0 0 1 0 10.25v-7.5A1.75 1.75 0 0 1 1.75 1h3a.75.75 0 0 1 0 1.5h-3a.25.25 0 0 0-.177.073ZM6.982 12a5.72 5.72 0 0 1-.765 2.5h3.566a5.72 5.72 0 0 1-.765-2.5H6.982Z"></path></svg> <span>Display lowest prices on the Steam store by <u>installing our extension</u></span></a>
</div>
</div>
</div>
</div>
<div class="container container-events">
<div class="row">
<div class="span12 sticky-year">
<h3>Upcoming events</h3>
<div class="sticky-year-buttons">
<a href="/sales/takeovers/" class="btn btn-outline" style="margin-right:6px">Takeovers</a>
<a href="/sales/newspopups/" class="btn btn-outline" style="margin-right:6px">Pop-ups</a>
<a href="/sales/history/all/" class="btn btn-outline">View all sale events</a>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/4145017/f1381c5c15060490e448f2a9ff767057cb449ac3.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/4145017/f1381c5c15060490e448f2a9ff767057cb449ac3.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3685685238933646707" target="_blank">Next Fest: February 2024</a>
<time class="future-sale-time" datetime="2024-02-05T18:00:00+00:00" title="5 February 2024 at 18:00:00 UTC">in 7 days</time>
</h4>
<div>5 â€” 12 February</div>
<div class="i muted">
Monday
</div>
<a href="https://discord.gg/steamdb?event=1150851172811419658" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/blog/1/95d3e755e0b80b42c21d1fda454e6b5aacd84e18.png)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/blog/1/95d3e755e0b80b42c21d1fda454e6b5aacd84e18.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3687933234119814601" target="_blank">Remote Play Together</a>
<time class="future-sale-time" datetime="2024-02-12T18:00:00+00:00" title="12 February 2024 at 18:00:00 UTC">in 14 days</time>
</h4>
<div>12 â€” 19 February</div>
<div class="i muted">
Monday
</div>
<a href="https://discord.gg/steamdb?event=1154148196998713344" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/blog/1/ecba77f2c7d0dd8216735fdd8f261c2b4a4eea1d.png)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/blog/1/ecba77f2c7d0dd8216735fdd8f261c2b4a4eea1d.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3687933234119814601" target="_blank">Dinos vs. Robots Fest</a>
<time class="future-sale-time" datetime="2024-02-26T18:00:00+00:00" title="26 February 2024 at 18:00:00 UTC">in 28 days</time>
</h4>
<div>26 February â€” 4 March</div>
<div class="i muted">
Monday
</div>
<a href="https://discord.gg/steamdb?event=1154148342297788446" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3687933234119814601" target="_blank">Spring Sale 2024</a>
<time class="future-sale-time" datetime="2024-03-14T17:00:00+00:00" title="14 March 2024 at 17:00:00 UTC">in 2 months</time>
</h4>
<div>14 â€” 21 March</div>
<div class="i muted">
Thursday
</div>
<a href="https://discord.gg/steamdb?event=1150851018398122075" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3687933234119814601" target="_blank">Deckbuilders Fest</a>
<time class="future-sale-time" datetime="2024-03-25T17:00:00+00:00" title="25 March 2024 at 17:00:00 UTC">in 2 months</time>
</h4>
<div>25 March â€” 1 April</div>
<div class="i muted">
Monday
</div>
<a href="https://discord.gg/steamdb?event=1154148443531526225" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3788151013299508612" target="_blank">FPS Fest</a>
<time class="future-sale-time" datetime="2024-04-15T17:00:00+00:00" title="15 April 2024 at 17:00:00 UTC">in 3 months</time>
</h4>
<div>15 â€” 22 April</div>
<div class="i muted">
Monday
</div>
<a href="https://discord.gg/steamdb?event=1183823990561701959" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3788151013299508612" target="_blank">Farming Fest</a>
<time class="future-sale-time" datetime="2024-04-29T17:00:00+00:00" title="29 April 2024 at 17:00:00 UTC">in 3 months</time>
</h4>
<div>29 April â€” 6 May</div>
<div class="i muted">
Monday
</div>
<a href="https://discord.gg/steamdb?event=1183824447841509396" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3788151013299508612" target="_blank">Endless Replayability Fest</a>
<time class="future-sale-time" datetime="2024-05-13T17:00:00+00:00" title="13 May 2024 at 17:00:00 UTC">in 4 months</time>
</h4>
<div>13 â€” 20 May</div>
<div class="i muted">
Monday
</div>
<a href="https://discord.gg/steamdb?event=1183824556146823189" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3788151013299508612" target="_blank">Open World Survival Crafting Fest</a>
<time class="future-sale-time" datetime="2024-05-27T17:00:00+00:00" title="27 May 2024 at 17:00:00 UTC">in 4 months</time>
</h4>
<div>27 May â€” 3 June</div>
<div class="i muted">
Monday
</div>
<a href="https://discord.gg/steamdb?event=1183824727085699203" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3788151013299508612" target="_blank">Next Fest: June 2024</a>
<time class="future-sale-time" datetime="2024-06-10T17:00:00+00:00" title="10 June 2024 at 17:00:00 UTC">in 4 months</time>
</h4>
<div>10 â€” 17 June</div>
<div class="i muted">
Monday
</div>
<a href="https://discord.gg/steamdb?event=1183824960687439963" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/4145017/38e29c16b5b2a3a50db2911ccdca69a84b3d6a33.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3788151013299508612" target="_blank">Summer Sale 2024</a>
<time class="future-sale-time" datetime="2024-06-27T17:00:00+00:00" title="27 June 2024 at 17:00:00 UTC">in 5 months</time>
</h4>
<div>27 June â€” 11 July</div>
<div class="i muted">
Thursday
</div>
<a href="https://discord.gg/steamdb?event=1183825329391947796" class="discord-event">
<svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-bell-fill" aria-hidden="true"><path d="M8 16c.9 0 1.7-.6 1.9-1.5.1-.3-.1-.5-.4-.5h-3c-.3 0-.5.2-.4.5.2.9 1 1.5 1.9 1.5ZM3 5c0-2.8 2.2-5 5-5s5 2.2 5 5v3l1.7 2.6c.2.2.3.5.3.8 0 .8-.7 1.5-1.5 1.5h-11c-.8.1-1.5-.6-1.5-1.4 0-.3.1-.6.3-.8L3 8.1V5Z"></path></svg> Follow event in Discord
</a>
</div>
</div>
</div>
<div class="panel panel-faq" style="margin-top:50px">
<div class="panel-heading d-flex">
<span class="flex-grow">Seasonal Steam Sales</span>
<a href="https://partner.steamgames.com/doc/marketing/discounts">Docs</a>
</div>
<p>Seasonal sales are created and managed by Steam. When selecting titles to fill these front-page promotional slots, Steam tries to be as fair as possible and select titles that customers are most interested in and will result in the most happy customers. While there aren't strict rules, as a base guideline Steam tends to focus on the top 10-20% selling games on Steam that are positively reviewed and have otherwise proven to be successful.</p>
<p>There are several sale events per year, which publishers and developers may have the chance to participate in. Some of these, like the Summer Sale or Winter Sale, are open to all publishers. Some may be only open to products of a certain category, like Anime or VR.</p>
</div>
<div class="panel panel-faq">
<div class="panel-heading d-flex">
<span class="flex-grow">Steam Themed Sale Events</span>
<a href="https://partner.steamgames.com/doc/marketing/upcoming_events/themed_sales">Docs</a>
</div>
<p>Steam regularly hosts themed sale events (typically designated as Fests) to showcase various genres and categories of games. Note that a discount is not required to participate in these events, though the event page will most prominently feature discounted games.</p>
<p>Each themed sale event has different eligibility criteria built around the core theme of the event, usually determined by the primary mechanic(s) of the game. For each event, Steam will automatically select and invite eligible games where the primary theme or mechanic of the game is in alignment with the theme of the sale. Partners usually have six weeks or more to register for the event and enter any desired discounts.</p>
</div>
<div class="panel panel-faq">
<div class="panel-heading d-flex">
<span class="flex-grow">Steam Next Fest</span>
<a href="https://partner.steamgames.com/doc/marketing/upcoming_events/nextfest">Docs</a>
</div>
<p>Steam Next Fest is a multi-day celebration where fans can try out demos, chat with developers, watch livestreams, and learn about upcoming games on Steam. For developers, Steam Next Fest is an opportunity to get early feedback from players and build an audience for a future launch on Steam. Steam typically hosts three editions of Steam Next Fest each year, which take place in February, June, and October.</p>
<p>Steam Fests are not sale events, as participating games are NOT required to go on sale. However, Fests usually do feature discounts.</p>
</div>
<div class="panel panel-faq">
<div class="panel-heading">Do you know when the next sale is and what discounts will it have?</div>
<p>Game publishers set the discounts, and they can choose any discount they want. You can look up any game on our site and look at its price history to determine which previous sales the game was discounted in. If a game appeared in all previous major sales, it's pretty safe to say it will be discounted again.</p>
<p>We do not know what specific games will be discounted in an upcoming sale, what prices will be, or what discounts they will have. Thematic sales are limited to games the sale is about.</p>
</div>
<a href="/donate/" class="armored-headcrab">
<span>SteamDB has been running ad-free since 2012.<br><b>Donate or contribute.</b></span>
<svg width="32" height="32" viewBox="0 0 16 16" class="octicon octicon-arrow-right" aria-hidden="true"><path d="M8.22 2.97a.75.75 0 0 1 1.06 0l4.25 4.25a.75.75 0 0 1 0 1.06l-4.25 4.25a.751.751 0 0 1-1.042-.018.751.751 0 0 1-.018-1.042l2.97-2.97H3.75a.75.75 0 0 1 0-1.5h7.44L8.22 4.03a.75.75 0 0 1 0-1.06Z"></path></svg>
</a>
<div class="row">
<div class="span12 sticky-year">
<h3>Year 2024</h3>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/41316928/30e8cbb6d033ce2cde553f14d0edd2d853f05929.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/41316928/30e8cbb6d033ce2cde553f14d0edd2d853f05929.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3687933234119814601" target="_blank">Pirates vs. Ninjas Fest</a>
</h4>
<div>22 â€” 29 January</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/e078052162aa0ec27ae8f523a1e5ec10a8e83a46.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/e078052162aa0ec27ae8f523a1e5ec10a8e83a46.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3901870710019684988" target="_blank">Capitalism and Economy Fest</a>
</h4>
<div>8 â€” 15 January</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
</div>
<div class="row">
<div class="span12 sticky-year">
<h3>Year 2023</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/d52d640338dc0ab64488f5d2824d6e8a1612fad6.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/d52d640338dc0ab64488f5d2824d6e8a1612fad6.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3877100446784190797" target="_blank">Winter Sale 2023</a>
</h4>
<div>21 December â€” 4 January 2024</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/fd54e11b3231375d2527c034b5666046be9315cb.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/fd54e11b3231375d2527c034b5666046be9315cb.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3895113407862688151" target="_blank">Steam VR Fest 2023</a>
</h4>
<div>4 â€” 11 December</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/313b58c9bda10bce760bd3c3f6dc635e1d90eeeb.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/313b58c9bda10bce760bd3c3f6dc635e1d90eeeb.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/app/593110/view/3823053915973576611" target="_blank">Autumn Sale 2023</a>
</h4>
<div>21 â€” 28 November</div>
<div class="i muted">
Tuesday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/44096841/88f8521b0f14f8e7d7e67a9cbbd5fadc3b493806.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/44096841/88f8521b0f14f8e7d7e67a9cbbd5fadc3b493806.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/2434320/announcements/detail/3709334211523789356" target="_blank">Halloween Sale 2023</a>
</h4>
<div>26 October â€” 2 November</div>
<div class="i muted">
Thursday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/798a61e6712ef47d6e961fb696838a2479cdcfa4.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/798a61e6712ef47d6e961fb696838a2479cdcfa4.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3711586646022775055" target="_blank">Next Fest: October 2023</a>
</h4>
<div>9 â€” 16 October</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/cc41da43a8f40aa25d36742cd501733cb586ce2c.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/cc41da43a8f40aa25d36742cd501733cb586ce2c.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3667672744266350605" target="_blank">SHMUP Fest</a>
</h4>
<div>25 September â€” 2 October</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/39049601/2dcc26864d22e743532967927ae1d3da55812a10.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/39049601/2dcc26864d22e743532967927ae1d3da55812a10.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3690192104503970270" target="_blank">Steam 20th Anniversary</a>
</h4>
<div>12 â€” 12 September</div>
<div class="i muted">
Tuesday, lasted 0 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/9682b9aa5345c8003ade8e372b9b3eb52337e4f1.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/9682b9aa5345c8003ade8e372b9b3eb52337e4f1.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3656413110959860073" target="_blank">Strategy Fest</a>
</h4>
<div>28 August â€” 4 September</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/34a5e4448f739707be7238d84a9d9fecf8e1aaff.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/34a5e4448f739707be7238d84a9d9fecf8e1aaff.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3656410664812213046" target="_blank">Visual Novel Fest</a>
</h4>
<div>7 â€” 14 August</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/0bbd120658b909fcb415a6857a48992ef46296f9.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/0bbd120658b909fcb415a6857a48992ef46296f9.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/app/593110/view/3690185668844106880" target="_blank">Stealth Fest</a>
</h4>
<div>24 â€” 31 July</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://clan.cloudflare.steamstatic.com/images/27766192/4b8392f48e9ec3beeca91cdc1ba7134e78ac236c.jpg)"></div>
<img class="sale-image" src="https://clan.cloudflare.steamstatic.com/images/27766192/4b8392f48e9ec3beeca91cdc1ba7134e78ac236c.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/app/593110/view/3682303734604033314" target="_blank">Summer Sale 2023</a>
</h4>
<div>29 June â€” 13 July</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4145017/88951c71a236817007550c5ee7f74932983c9332.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4145017/88951c71a236817007550c5ee7f74932983c9332.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3648510249385323491" target="_blank">Next Fest: June 2023</a>
</h4>
<div>19 â€” 26 June</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/204e7f746f1e425f059d0a4b9f9131668479ac58.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/204e7f746f1e425f059d0a4b9f9131668479ac58.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/app/593110/view/3719450371593620107" target="_blank">Sports Fest</a>
</h4>
<div>15 â€” 22 May</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/3a761dedb128ed41ff90d292847f4255a9e9e28b.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/3a761dedb128ed41ff90d292847f4255a9e9e28b.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3719450371590731903" target="_blank">Puzzle Fest</a>
</h4>
<div>24 April â€” 1 May</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/7b701483f388e11c9919f898deaa11b22f10ed65.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/7b701483f388e11c9919f898deaa11b22f10ed65.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3694679839286985809" target="_blank">Spring Sale 2023</a>
</h4>
<div>16 â€” 23 March</div>
<div class="i muted">
Thursday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/5d27d3986d7251507cd9d99942f3a8b4c78b875a.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/5d27d3986d7251507cd9d99942f3a8b4c78b875a.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3682291689319062483" target="_blank">Mystery Fest</a>
</h4>
<div>20 â€” 27 February</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/21f901a028ac1d94eef7b46e6ce31ad0ad024863.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/21f901a028ac1d94eef7b46e6ce31ad0ad024863.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3638380325135497082" target="_blank">Next Fest: February 2023</a>
</h4>
<div>6 â€” 13 February</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/658f357a266a308036407f0cd7aa81019708cd9d.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/658f357a266a308036407f0cd7aa81019708cd9d.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/app/593110/view/3616985051989391382" target="_blank">Base Builder Fest</a>
</h4>
<div>23 â€” 30 January</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39769908/2c5bfb1dece7b917a96943119cfb00962bb47a94.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39769908/2c5bfb1dece7b917a96943119cfb00962bb47a94.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/group/39769908/view/6390075983611996033" target="_blank">Lunar New Year Sale 2023</a>
</h4>
<div>18 â€” 27 January</div>
<div class="i muted">
Wednesday, lasted 9 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2022</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/c917286aecad32045f0e06d047f7d7d6c78369f7.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/c917286aecad32045f0e06d047f7d7d6c78369f7.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3653012651448289509" target="_blank">Winter Sale 2022</a>
</h4>
<div>22 December â€” 5 January 2023</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/d0e635d2c233e7c0e780a666179b3130af1bc715.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/d0e635d2c233e7c0e780a666179b3130af1bc715.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3388424091280374210" target="_blank">Autumn Sale 2022</a>
</h4>
<div>22 â€” 29 November</div>
<div class="i muted">
Tuesday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/faaf88fa7ab3212e3d60adcadc434c7985ec81e0.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/faaf88fa7ab3212e3d60adcadc434c7985ec81e0.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3329873512170602587" target="_blank">Halloween Sale 2022</a>
</h4>
<div>25 October â€” 1 November</div>
<div class="i muted">
Tuesday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/fac1fba7f105be3b6fd1fd1f07dcb87a0d1abb7c.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/fac1fba7f105be3b6fd1fd1f07dcb87a0d1abb7c.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3305103170305173927" target="_blank">Next Fest: October 2022</a>
</h4>
<div>3 â€” 10 October</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/7a96241dbe13ab18376b9a57dd53d1d8a7417e68.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/7a96241dbe13ab18376b9a57dd53d1d8a7417e68.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3228538806406693494" target="_blank">Steam Bash Bash</a>
</h4>
<div>19 â€” 26 September</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/2029fd56d27395a36bff9a912fdb2483d90609b9.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/2029fd56d27395a36bff9a912fdb2483d90609b9.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3361392458123361051" target="_blank">Steam Survival Fest 2022</a>
</h4>
<div>1 â€” 8 August</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/41316928/7a7aa9dae57322d190595c48eaf30bc4111baf25.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/41316928/7a7aa9dae57322d190595c48eaf30bc4111baf25.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3335493590032031321" target="_blank">Steam VR Fest 2022</a>
</h4>
<div>18 â€” 25 July</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/3cf580e0be5cbf0086861873b38f29ddc47c8155.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/3cf580e0be5cbf0086861873b38f29ddc47c8155.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3335493590017083918" target="_blank">Summer Sale 2022</a>
</h4>
<div>23 June â€” 7 July</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/be146d0556a20a1db048dc29773291fcf2622d0c.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/be146d0556a20a1db048dc29773291fcf2622d0c.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/app/593110/view/3341121821003187514" target="_blank">Next Fest: June 2022</a>
</h4>
<div>13 â€” 20 June</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/41316928/c12276746776f2205e36561c96c40c5f38477fdc.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/41316928/c12276746776f2205e36561c96c40c5f38477fdc.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3180116240852632901" target="_blank">Racing Fest 2022</a>
</h4>
<div>23 â€” 30 May</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/1e64d836d0d7f0a15f33ff568688ce99a1642106.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/1e64d836d0d7f0a15f33ff568688ce99a1642106.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3175610095143798951" target="_blank">Going Rogue: A Festival of Persistence</a>
</h4>
<div>2 â€” 9 May</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/42048709/306975a1791b1d1bf44fdb0af7c286d023e211ca.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/42048709/306975a1791b1d1bf44fdb0af7c286d023e211ca.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/group/42048709/view/4689941763337468111" target="_blank">Simfest: Hobby Edition</a>
</h4>
<div>28 March â€” 4 April</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/41316928/001dafc366b108c1dbb998d8186ec3f2b9c55a78.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/41316928/001dafc366b108c1dbb998d8186ec3f2b9c55a78.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3091163163109910645" target="_blank">JRPG Sale</a>
</h4>
<div>14 â€” 21 March</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4145017/cc120bd39922297d8bc38e30c8c010d852fe05c3.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4145017/cc120bd39922297d8bc38e30c8c010d852fe05c3.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/4330775883772405894" target="_blank">Remote Play Together 2022</a>
</h4>
<div>28 February â€” 7 March</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/c05b8ab5d6d66dfe466b3a451efd740302d051b0.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/c05b8ab5d6d66dfe466b3a451efd740302d051b0.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/store_promos/announcements/detail/3119309392144252575" target="_blank">Next Fest: February 2022</a>
</h4>
<div>21 â€” 28 February</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/344da74ae6b2d3ba312d26d314e251e7f8a88872.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/344da74ae6b2d3ba312d26d314e251e7f8a88872.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3100165379962854007" target="_blank">Lunar New Year Sale 2022</a>
</h4>
<div>27 January â€” 3 February</div>
<div class="i muted">
Thursday, lasted 7 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2021</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/steamworks_docs/english/SteamSocialMediaShare_EN.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/steamworks_docs/english/SteamSocialMediaShare_EN.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://partner.steamgames.com/doc/marketing/upcoming_events" target="_blank">Winter Sale 2021</a>
</h4>
<div>22 December â€” 5 January 2022</div>
<div class="i muted">
Wednesday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/8bbf1d789f61a273fd0356e0f2433c56f350ecdb.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/8bbf1d789f61a273fd0356e0f2433c56f350ecdb.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/4555948886325351153" target="_blank">Autumn Sale 2021</a>
</h4>
<div>24 November â€” 1 December</div>
<div class="i muted">
Wednesday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/be2e190ebb4c62a3aa0a4d057f4d430340e18702.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/be2e190ebb4c62a3aa0a4d057f4d430340e18702.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3021345332649804118" target="_blank">Halloween Sale 2021</a>
</h4>
<div>28 October â€” 1 November</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/25c9defc765a4a5d8e2372d6216fff01b9ef8a56.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/25c9defc765a4a5d8e2372d6216fff01b9ef8a56.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/app/593110/view/4625752147812909151" target="_blank">Digital Tabletop Fest 2021</a>
</h4>
<div>21 â€” 25 October</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/d700aa6ad365d2823f78365629ebeaa77529e6b4.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/d700aa6ad365d2823f78365629ebeaa77529e6b4.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/steamworks/announcements/detail/3001069789761631205" target="_blank">Next Fest: October 2021</a>
</h4>
<div>1 â€” 7 October</div>
<div class="i muted">
Friday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/f85ad27e40fb52ab7a0977efc17892c2eb67272e.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/f85ad27e40fb52ab7a0977efc17892c2eb67272e.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/2981927408950091621" target="_blank">Summer Sale 2021</a>
</h4>
<div>24 June â€” 8 July</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/a984f02fdf0c0318d42c453110403297cf6f9ef4.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/a984f02fdf0c0318d42c453110403297cf6f9ef4.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/4444470938463675954" target="_blank">Next Fest: June 2021</a>
</h4>
<div>16 â€” 22 June</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4145017/ce2e4f63db71d710db21921762bd445da1332e59.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4145017/ce2e4f63db71d710db21921762bd445da1332e59.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3017953668435588364" target="_blank">Open World Sale</a>
</h4>
<div>27 â€” 31 May</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/8e31d6015f0100a6fc13dd723b02c3912e7b8809.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/8e31d6015f0100a6fc13dd723b02c3912e7b8809.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3035965529921562666" target="_blank">Golden Week Sale 2021</a>
</h4>
<div>29 April â€” 6 May</div>
<div class="i muted">
Thursday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/8c9e6b900176c8952d23ef341b29a6a80ee2b872.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/8c9e6b900176c8952d23ef341b29a6a80ee2b872.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/gid/103582791468571009/announcements/detail/3047218819063335508" target="_blank">Remote Play Together 2021</a>
</h4>
<div>25 â€” 29 March</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/326ee439d007ffcbac058c86787321c1bfd737ce.jpg?t=1613067179)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/326ee439d007ffcbac058c86787321c1bfd737ce.jpg?t=1613067179" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3016818255535621678" target="_blank">Lunar New Year Sale 2021</a>
</h4>
<div>11 â€” 15 February</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/7b7b71998d0992dd1f062f174511c050cb35441b.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/39049601/7b7b71998d0992dd1f062f174511c050cb35441b.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/store_promos/announcements/detail/2909851422942194402" target="_blank">Game Festival: February 2021</a>
</h4>
<div>3 â€” 9 February</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2020</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/97e0120731f8938ae061a78d377a92838a927782.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/97e0120731f8938ae061a78d377a92838a927782.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/2904223191309839604" target="_blank">Winter Sale 2020</a>
</h4>
<div>22 December â€” 5 January 2021</div>
<div class="i muted">
Tuesday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/7732853f0cd05321f8db812c8e5d35f1ede98780.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/7732853f0cd05321f8db812c8e5d35f1ede98780.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/2914353120901701711" target="_blank">Autumn Sale 2020</a>
</h4>
<div>25 November â€” 1 December</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/cc69a448f527f9c8d523dd6a8dadee67e4a1c02c.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/cc69a448f527f9c8d523dd6a8dadee67e4a1c02c.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/5092967634924292655" target="_blank">Fighting Game Sale</a>
</h4>
<div>12 â€” 16 November</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/ce414c53e810194624a058ebffd8c480427f0633.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/ce414c53e810194624a058ebffd8c480427f0633.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/5092967634924390947" target="_blank">City Builder Sale</a>
</h4>
<div>3 â€” 10 November</div>
<div class="i muted">
Tuesday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/a9e31e0f2a9b126b7fb58a8b316335dd4cf35b25.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/a9e31e0f2a9b126b7fb58a8b316335dd4cf35b25.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/2924484951654815100" target="_blank">Halloween Sale 2020</a>
</h4>
<div>29 October â€” 2 November</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/38511516/48431766aba865a2fb13733615b029a738ca170b.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/38511516/48431766aba865a2fb13733615b029a738ca170b.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/2769108226045128183" target="_blank">Digital Tabletop Fest</a>
</h4>
<div>21 â€” 26 October</div>
<div class="i muted">
Wednesday, lasted 5 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/d74fc7258e28650c0ae8872a7de6d16178d0e59d.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/d74fc7258e28650c0ae8872a7de6d16178d0e59d.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/2722942525864254967" target="_blank">Game Festival: Autumn Edition</a>
</h4>
<div>7 â€” 13 October</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/dd88f1201e9dcb6f03bf74ed11c3ba6ee3582cc0.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/dd88f1201e9dcb6f03bf74ed11c3ba6ee3582cc0.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/gid/103582791429521412/announcements/detail/2874940914770617768" target="_blank">Tokyo Game Show 2020</a>
</h4>
<div>24 â€” 28 September</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/adc3c2ede743062fc0699039ea812843e90c5eca.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/adc3c2ede743062fc0699039ea812843e90c5eca.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/2719562923901875829" target="_blank">Pirate Sale</a>
</h4>
<div>19 â€” 21 September</div>
<div class="i muted">
Saturday, lasted 2 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/32943613/8eb82dba3b5fa7e828c1c5e79edc9e8cd74c09ce.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/32943613/8eb82dba3b5fa7e828c1c5e79edc9e8cd74c09ce.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/newshub/group/32943613/view/2543919999013676243" target="_blank">Swiss Developer Sale</a>
</h4>
<div>27 July â€” 3 August</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/717f099e3d21f1f13ee6711f3124e45aaddc7a9f.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/717f099e3d21f1f13ee6711f3124e45aaddc7a9f.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/2445966074571282888" target="_blank">Summer Sale 2020</a>
</h4>
<div>25 June â€” 9 July</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/0c6d1412166e43ba26faea2724034e2cb00d0f52.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/0c6d1412166e43ba26faea2724034e2cb00d0f52.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/4534507336203365606" target="_blank">Game Festival: Summer Edition</a>
</h4>
<div>16 â€” 22 June</div>
<div class="i muted">
Tuesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/03161c2a3850f924fd62849ffe560a69638d4a66.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/03161c2a3850f924fd62849ffe560a69638d4a66.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/2173495758984412525" target="_blank">Spring Cleaning Event 2020</a>
</h4>
<div>21 â€” 28 May</div>
<div class="i muted">
Thursday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/06eafd98962181c31b03e550777889049b5c57f7.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/06eafd98962181c31b03e550777889049b5c57f7.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/1795187684692798629" target="_blank">Golden Week Sale 2020</a>
</h4>
<div>30 April â€” 6 May</div>
<div class="i muted">
Thursday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/ae3bb9ba84bce6849533d734525054f60433d9fd.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/ae3bb9ba84bce6849533d734525054f60433d9fd.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/1696107224077118345" target="_blank">Game Festival: Spring Edition</a>
</h4>
<div>18 â€” 23 March</div>
<div class="i muted">
Wednesday, lasted 5 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/1248000/capsule_616x353.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/1248000/capsule_616x353.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/1688220857098505776" target="_blank">Soundtrack Sale</a>
</h4>
<div>11 â€” 14 February</div>
<div class="i muted">
Tuesday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/7d927da5e19aabcfe254593ab52f6492847236b3.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/7d927da5e19aabcfe254593ab52f6492847236b3.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/1706236518822639430" target="_blank">Lunar New Year Sale 2020</a>
</h4>
<div>23 â€” 27 January</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2019</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/f5a67a84d6a45b9d89bfbcfe537cafe4a1fed4cd.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/f5a67a84d6a45b9d89bfbcfe537cafe4a1fed4cd.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3180035790689045584" target="_blank">Winter Sale 2019</a>
</h4>
<div>19 December â€” 2 January 2020</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/51f9c73ac72761c7d0c5aa7d9ba764bc310c9279.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/4/51f9c73ac72761c7d0c5aa7d9ba764bc310c9279.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/1610529315704894098" target="_blank">Game Awards 2019</a>
</h4>
<div>12 â€” 13 December</div>
<div class="i muted">
Thursday, lasted 1 day
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/4fd43bd2f972a8bd5a15a4fb2a78300d3c706372.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/4fd43bd2f972a8bd5a15a4fb2a78300d3c706372.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/3605623508392801946" target="_blank">Autumn Sale 2019</a>
</h4>
<div>26 November â€” 3 December</div>
<div class="i muted">
Tuesday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/355abc07a082b58c184d4c7f9303de8f68d61aff.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/355abc07a082b58c184d4c7f9303de8f68d61aff.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/1646555578396366598" target="_blank">Remote Play Together 2019</a>
</h4>
<div>20 â€” 25 November</div>
<div class="i muted">
Wednesday, lasted 5 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/1192850/capsule_616x353.jpg?t=1573329069)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/1192850/capsule_616x353.jpg?t=1573329069" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/groups/Valve/announcements/detail/1590261853010354336" target="_blank">Singles' Day Sale</a>
</h4>
<div>10 â€” 12 November</div>
<div class="i muted">
Sunday, lasted 2 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2019_halloween.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2019_halloween.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/1188864251785433090" target="_blank">Halloween Sale 2019</a>
</h4>
<div>28 October â€” 1 November</div>
<div class="i muted">
Monday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_lgbtq-sale/2d453aa30870f3a58bd8bdf4/share_image_english.jpg?t=1570818456)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_lgbtq-sale/2d453aa30870f3a58bd8bdf4/share_image_english.jpg?t=1570818456" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/1182718985990737920" target="_blank">LGBTQ+ Games Sale</a>
</h4>
<div>11 â€” 15 October</div>
<div class="i muted">
Friday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/52498/capsule_lg.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/52498/capsule_lg.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/52498" target="_blank">Space Exploration Sale 2019</a>
</h4>
<div>18 â€” 21 July</div>
<div class="i muted">
Thursday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2019_summer.png)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2019_summer.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/1616144864293404894" target="_blank">Summer Sale 2019 (Grand Prix)</a>
</h4>
<div>25 June â€” 9 July</div>
<div class="i muted">
Tuesday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/51494/capsule_lg.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/51494/capsule_lg.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/51494" target="_blank">World Environment Day Sale 2019</a>
</h4>
<div>4 â€” 7 June</div>
<div class="i muted">
Tuesday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2019_springcleaning.png)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2019_springcleaning.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/1622897728543594121" target="_blank">Spring Cleaning Event 2019</a>
</h4>
<div>24 â€” 28 May</div>
<div class="i muted">
Friday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2019_goldenweek.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2019_goldenweek.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/1123274564572573696" target="_blank">Golden Week Sale 2019</a>
</h4>
<div>29 April â€” 6 May</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2019_bafta.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2019_bafta.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/1113852479844769792" target="_blank">BAFTA Sale 2019</a>
</h4>
<div>4 â€” 13 April</div>
<div class="i muted">
Thursday, lasted 9 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2019_texas.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2019_texas.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/1106295257061187585" target="_blank">SXSW Texas Sale 2019</a>
</h4>
<div>14 â€” 17 March</div>
<div class="i muted">
Thursday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/0bc3fe5a06c19f9f9a50e5c8ecbf761342272525.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/0bc3fe5a06c19f9f9a50e5c8ecbf761342272525.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/1750114346364440017" target="_blank">Lunar New Year Sale 2019</a>
</h4>
<div>4 â€” 11 February</div>
<div class="i muted">
Monday, lasted 7 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2018</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/b5c9a94b98c8156cb4df3bfb1796eacc9dc0eabc.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/27766192/b5c9a94b98c8156cb4df3bfb1796eacc9dc0eabc.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://steamcommunity.com/games/593110/announcements/detail/2796070940830444045" target="_blank">Winter Sale 2018 (Extremely Cozy Cottage of Surprises)</a>
</h4>
<div>20 December â€” 3 January 2019</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/46135/news_post.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/46135/news_post.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/46126" target="_blank">Autumn Sale 2018 (Steam Awards Nominations)</a>
</h4>
<div>21 â€” 27 November</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2018_norway.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2018_norway.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/1062453801230028800" target="_blank">Games from Norway Sale 2018</a>
</h4>
<div>13 â€” 16 November</div>
<div class="i muted">
Tuesday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2018_poland.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2018_poland.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/1059942464138682368" target="_blank">100 Years of Polish Independance Sale 2018</a>
</h4>
<div>7 â€” 8 November</div>
<div class="i muted">
Wednesday, lasted 1 day
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2018_halloween.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2018_halloween.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/1056969343571976196" target="_blank">Halloween Sale 2018</a>
</h4>
<div>29 October â€” 1 November</div>
<div class="i muted">
Monday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2018_girls.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2018_girls.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/1050430836779167745" target="_blank">Day of the Girl Sale 2018</a>
</h4>
<div>11 â€” 12 October</div>
<div class="i muted">
Thursday, lasted 1 day
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/41369/Logo_Banner.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/41369/Logo_Banner.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/41369" target="_blank">Summer Sale 2018</a>
</h4>
<div>21 June â€” 5 July</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/40462/News_Header.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/40462/News_Header.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/40462" target="_blank">Spring Cleaning Event 2018</a>
</h4>
<div>24 â€” 28 May</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_gamesfromdenmark/9d0611d62f7b32185be73ea2/page_bg_english.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_gamesfromdenmark/9d0611d62f7b32185be73ea2/page_bg_english.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/39433" target="_blank">Games from Denmark Sale 2018</a>
</h4>
<div>24 â€” 27 April</div>
<div class="i muted">
Tuesday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/39076/Large_cap.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/39076/Large_cap.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/39076" target="_blank">VR Spring Sale 2018</a>
</h4>
<div>19 â€” 23 April</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/38343/spotlight.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/38343/spotlight.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/38343" target="_blank">GDC Film Festival Sale 2018</a>
</h4>
<div>19 â€” 21 March</div>
<div class="i muted">
Monday, lasted 2 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/3953098/14b7941eb7ce85be3fde34c61a16c3de814abb6f.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steamcommunity/public/images/clans/3953098/14b7941eb7ce85be3fde34c61a16c3de814abb6f.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/37184" target="_blank">Lunar New Year Sale 2018</a>
</h4>
<div>15 â€” 19 February</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2017</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/35783/Day1_ChoicesMatter.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/35783/Day1_ChoicesMatter.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/35783" target="_blank">Winter Sale 2017</a>
</h4>
<div>21 December â€” 4 January 2018</div>
<div class="i muted">
Thursday, lasted 14 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/35237/large_capsule.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/35237/large_capsule.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/35237" target="_blank">Finland Anniversary Sale 2017</a>
</h4>
<div>5 â€” 8 December</div>
<div class="i muted">
Tuesday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/34820/news_post_1.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/34820/news_post_1.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/34820" target="_blank">Autumn Sale 2017 (Steam Awards Nominations)</a>
</h4>
<div>22 â€” 28 November</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/33893/news_post1.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/33893/news_post1.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/33893" target="_blank">Halloween Sale 2017</a>
</h4>
<div>26 October â€” 1 November</div>
<div class="i muted">
Thursday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/30293/news_post_5.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/30293/news_post_5.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/30293" target="_blank">Summer Sale 2017</a>
</h4>
<div>22 June â€” 5 July</div>
<div class="i muted">
Thursday, lasted 13 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/28652/Main-Cap.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/28652/Main-Cap.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/28652" target="_blank">VR Anniversary Sale 2017</a>
</h4>
<div>5 â€” 9 April</div>
<div class="i muted">
Wednesday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/28487/large_cap.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/28487/large_cap.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/28487" target="_blank">Anime Sale 2017</a>
</h4>
<div>30 March â€” 3 April</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2016</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/26499/newspostimage.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/26499/newspostimage.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/26499" target="_blank">Winter Sale 2016</a>
</h4>
<div>22 December â€” 2 January 2017</div>
<div class="i muted">
Thursday, lasted 11 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/25825/SteamAutumnSale1.PNG)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/25825/SteamAutumnSale1.PNG" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/25825" target="_blank">Autumn Sale 2016 (Steam Awards Nominations)</a>
</h4>
<div>23 â€” 29 November</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/25262/news_post.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/25262/news_post.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/25262" target="_blank">Halloween Sale 2016</a>
</h4>
<div>28 October â€” 1 November</div>
<div class="i muted">
Friday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/23161/news_post.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/23161/news_post.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/23161" target="_blank">VR Weekend Sale 2016</a>
</h4>
<div>28 â€” 31 July</div>
<div class="i muted">
Thursday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2016_summer.png)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2016_summer.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/22603" target="_blank">Summer Sale 2016 (Picnic)</a>
</h4>
<div>23 June â€” 4 July</div>
<div class="i muted">
Thursday, lasted 11 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/489420/capsule_467x181.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/489420/capsule_467x181.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/22313" target="_blank">Soccer Sale 2016</a>
</h4>
<div>9 â€” 13 June</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/21582/spotlight.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/21582/spotlight.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/21582" target="_blank">Anime Sale 2016</a>
</h4>
<div>28 April â€” 1 May</div>
<div class="i muted">
Thursday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/20313/newspost.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/20313/newspost.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/20313" target="_blank">Lunar New Year Sale 2016</a>
</h4>
<div>5 â€” 12 February</div>
<div class="i muted">
Friday, lasted 7 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2015</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_winter2015_daily/148e7f4d9edeb0443fbeede7/page_bg_english.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_winter2015_daily/148e7f4d9edeb0443fbeede7/page_bg_english.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/19813" target="_blank">Winter Sale 2015 (Gingerbread Jake In: North Pole Noir)</a>
</h4>
<div>22 December â€” 4 January 2016</div>
<div class="i muted">
Tuesday, lasted 13 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/15094/newspostday2.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/15094/newspostday2.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/19452" target="_blank">Autumn Sale 2015 (Exploration)</a>
</h4>
<div>26 November â€” 2 December</div>
<div class="i muted">
Thursday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2015_halloween.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2015_halloween.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/660153014116708352" target="_blank">Halloween Sale 2015</a>
</h4>
<div>29 October â€” 2 November</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_stealth_sale/a18a26ef995ba46c787fa550/page_bg_english.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_stealth_sale/a18a26ef995ba46c787fa550/page_bg_english.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/18798" target="_blank">Stealth Sale 2015</a>
</h4>
<div>13 â€” 16 October</div>
<div class="i muted">
Tuesday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/17207/News-Day1.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/17207/News-Day1.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/17207" target="_blank">Summer Sale 2015 (Monster Summer)</a>
</h4>
<div>11 â€” 21 June</div>
<div class="i muted">
Thursday, lasted 10 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/16505/header.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/16505/header.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/16505" target="_blank">Rogue-like Games Sale 2015</a>
</h4>
<div>23 â€” 27 April</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2015_cyberpunk.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2015_cyberpunk.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/16114" target="_blank">Cyberpunk Sale 2015</a>
</h4>
<div>17 â€” 20 March</div>
<div class="i muted">
Tuesday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/351180/capsule_467x181.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/351180/capsule_467x181.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/15822" target="_blank">Co-op Sale 2015 (Valentine's Day)</a>
</h4>
<div>14 â€” 16 February</div>
<div class="i muted">
Saturday, lasted 2 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2015_mystery.png)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2015_mystery.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://twitter.com/steam_games/status/565211141970751488" target="_blank">Mystery Midweek Sale 2015</a>
</h4>
<div>10 â€” 13 February</div>
<div class="i muted">
Tuesday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/15418/message_image.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/15418/message_image.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/15418" target="_blank">Anime Sale 2015</a>
</h4>
<div>8 â€” 12 January</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2014</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/15308/winter_promo_img.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/15308/winter_promo_img.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/15308" target="_blank">Winter Sale 2014 (Holiday Auction)</a>
</h4>
<div>18 December â€” 2 January 2015</div>
<div class="i muted">
Thursday, lasted 15 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/15081/newspost-day1.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/15081/newspost-day1.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/15081" target="_blank">Autumn Sale 2014 (Exploration)</a>
</h4>
<div>26 November â€” 2 December</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/14819/news_capsule_600.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/14819/news_capsule_600.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/14819" target="_blank">Halloween Sale 2014</a>
</h4>
<div>30 October â€” 3 November</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/323950/header.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/323950/header.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/14422" target="_blank">French Indie Sale 2014</a>
</h4>
<div>17 â€” 19 September</div>
<div class="i muted">
Wednesday, lasted 2 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_japaneseindiefestival/4d5e8a4e/page_bg_english.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_japaneseindiefestival/4d5e8a4e/page_bg_english.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/14178" target="_blank">Japanese Indie Festival Sale 2014</a>
</h4>
<div>16 â€” 17 August</div>
<div class="i muted">
Saturday, lasted 1 day
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/312490/capsule_467x181.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/312490/capsule_467x181.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/13735" target="_blank">Canada Day Sale 2014</a>
</h4>
<div>1 â€” 3 July</div>
<div class="i muted">
Tuesday, lasted 2 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/13702/SummerSale_Newspost.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/13702/SummerSale_Newspost.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/13702" target="_blank">Summer Sale 2014 (Summer Adventure)</a>
</h4>
<div>19 â€” 30 June</div>
<div class="i muted">
Thursday, lasted 11 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/281710/capsule_467x181.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/281710/capsule_467x181.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/12413" target="_blank">Valentine's Day 2014</a>
</h4>
<div>14 â€” 15 February</div>
<div class="i muted">
Friday, lasted 1 day
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2013</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/12138/NewsPost.PNG)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/12138/NewsPost.PNG" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/12138" target="_blank">Winter Sale 2013 (Snow Globe)</a>
</h4>
<div>19 December â€” 3 January 2014</div>
<div class="i muted">
Thursday, lasted 15 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/11961/News_Post.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/11961/News_Post.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/11961" target="_blank">Autumn Sale 2013</a>
</h4>
<div>27 November â€” 3 December</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/11732/header_NewsPost.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/11732/header_NewsPost.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/11732" target="_blank">Halloween Sale 2013</a>
</h4>
<div>29 October â€” 1 November</div>
<div class="i muted">
Tuesday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/11359/News_Post.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/11359/News_Post.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/11359" target="_blank">Greenlight One Year Anniversary Sale 2013</a>
</h4>
<div>29 August â€” 2 September</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/11069/News_Post_Day1-Bioshock.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/11069/News_Post_Day1-Bioshock.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/11069" target="_blank">Summer Sale 2013</a>
</h4>
<div>11 â€” 22 July</div>
<div class="i muted">
Thursday, lasted 11 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/10190/capsule_lg.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/10190/capsule_lg.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/10190" target="_blank">Spring Indie Sale 2013</a>
</h4>
<div>21 â€” 29 March</div>
<div class="i muted">
Thursday, lasted 8 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2012</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/9635/day_1.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/9635/day_1.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/9635" target="_blank">Winter Sale 2012</a>
</h4>
<div>20 December â€” 5 January 2013</div>
<div class="i muted">
Thursday, lasted 16 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/9428/News_post_day1.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/9428/News_post_day1.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/9428" target="_blank">Autumn Sale 2012</a>
</h4>
<div>21 â€” 26 November</div>
<div class="i muted">
Wednesday, lasted 5 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/news/9225/capsule_lg.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/news/9225/capsule_lg.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/9225" target="_blank">Halloween Sale 2012</a>
</h4>
<div>29 October â€” 1 November</div>
<div class="i muted">
Monday, lasted 3 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/630/extras/summer_sale_capsule2.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/630/extras/summer_sale_capsule2.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/8388" target="_blank">Summer Sale 2012</a>
</h4>
<div>12 â€” 23 July</div>
<div class="i muted">
Thursday, lasted 11 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2011</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/620/extras/News_GiftPile.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/620/extras/News_GiftPile.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/7040" target="_blank">Winter Sale 2011 (Great Gift Pile)</a>
</h4>
<div>19 December â€” 1 January 2012</div>
<div class="i muted">
Monday, lasted 13 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/620/extras/header_day1.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/620/extras/header_day1.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/6844" target="_blank">Autumn Sale 2011</a>
</h4>
<div>23 â€” 27 November</div>
<div class="i muted">
Wednesday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/550/extras/capsule_lg.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/550/extras/capsule_lg.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/6618" target="_blank">Halloween Sale 2011</a>
</h4>
<div>27 â€” 31 October</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/31290/extras/release__scene01.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/31290/extras/release__scene01.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/5761" target="_blank">Summer Sale 2011 (Summer Camp)</a>
</h4>
<div>30 June â€” 10 July</div>
<div class="i muted">
Thursday, lasted 10 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2010</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_holiday2010_weeklong/97f089ae/treasurehunt_english.png)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_holiday2010_weeklong/97f089ae/treasurehunt_english.png" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/4816" target="_blank">Winter Sale 2010 (Treasure Hunt)</a>
</h4>
<div>20 December â€” 2 January 2011</div>
<div class="i muted">
Monday, lasted 13 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_giveandget_daily/f91c2ab2/comic_english.png?)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_giveandget_daily/f91c2ab2/comic_english.png?" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/4695" target="_blank">Autumn Sale 2010 (Give &amp; Get)</a>
</h4>
<div>24 â€” 30 November</div>
<div class="i muted">
Wednesday, lasted 6 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2010_halloween.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2010_halloween.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/4574" target="_blank">Halloween Sale 2010</a>
</h4>
<div>28 October â€” 1 November</div>
<div class="i muted">
Thursday, lasted 4 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_summer/b514eff4/swim_news_english.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/clusters/sale_summer/b514eff4/swim_news_english.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/3985" target="_blank">Summer Sale 2010 (Perils of Summer)</a>
</h4>
<div>24 June â€” 4 July</div>
<div class="i muted">
Thursday, lasted 10 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2009</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2009_halloween.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2009_halloween.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/3285" target="_blank">Winter Sale 2009</a>
</h4>
<div>22 December â€” 3 January 2010</div>
<div class="i muted">
Tuesday, lasted 12 days
</div>
</div>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://cdn.cloudflare.steamstatic.com/steam/apps/901180/header.jpg)"></div>
<img class="sale-image" src="https://cdn.cloudflare.steamstatic.com/steam/apps/901180/header.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/3163" target="_blank">Autumn Sale 2009</a>
</h4>
<div>25 â€” 30 November</div>
<div class="i muted">
Wednesday, lasted 5 days
</div>
</div>
</div>
<div class="span4 panel-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2009_indie.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2009_indie.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/2366" target="_blank">Indie Games Sale 2009</a>
</h4>
<div>20 â€” 29 March</div>
<div class="i muted">
Friday, lasted 9 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2008</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2008_holiday.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2008_holiday.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://store.steampowered.com/news/2127" target="_blank">Winter Sale 2008</a>
</h4>
<div>26 December â€” 2 January 2009</div>
<div class="i muted">
Friday, lasted 7 days
</div>
</div>
</div>
</div>
<div class="row hidden-event-row">
<div class="span12 sticky-year">
<h3>Year 2007</h3>
</div>
<div class="span4 panel-sale major-sale">
<div class="sale-image-bg" style="background-image:url(https://steamdb.info/static/img/sales/2007_holiday.jpg)"></div>
<img class="sale-image" src="https://steamdb.info/static/img/sales/2007_holiday.jpg" loading="lazy" alt="">
<div class="panel-body">
<h4 class="panel-sale-name">
<a href="https://web.archive.org/web/20071228151102/http://www.steampowered.com/v/index.php?area=news&amp;id=1382&amp;cc=US" target="_blank">Winter Sale 2007</a>
</h4>
<div>24 December â€” 1 January 2008</div>
<div class="i muted">
Monday, lasted 8 days
</div>
</div>
</div>
</div>
<div class="row">
<div class="span12 text-center">
<button class="btn" type="button" id="js-view-hidden-events" hidden="">View sale events for previous years</button>
</div>
</div>
</div>
<script nonce="" integrity="sha512-RK9HDvLjuaXQl6r+ZYQtUtoE/FYMf1iPtBRb5BiZqGlKDb8raWz8reUDbdADi6cXhACtXjbbnefuOxBYUFfW6w==" src="/static/js/timeago.js?v=44af470ef2e3b9a5"></script>
<script nonce="">
(function()
{
document.getElementById( 'js-view-hidden-events' ).addEventListener( 'click', function()
{
this.hidden = true;
document.querySelectorAll( '.hidden-event-row' ).forEach( ( el ) =>
{
el.hidden = false;
} );
} );
TimeAgo.initAll( '.future-sale-time' );
}());
</script>
</div>
<a href="/donate/" class="armored-headcrab">
<span>SteamDB has been running ad-free since 2012.<br><b>Donate or contribute.</b></span>
<svg width="32" height="32" viewBox="0 0 16 16" class="octicon octicon-arrow-right" aria-hidden="true"><path d="M8.22 2.97a.75.75 0 0 1 1.06 0l4.25 4.25a.75.75 0 0 1 0 1.06l-4.25 4.25a.751.751 0 0 1-1.042-.018.751.751 0 0 1-.018-1.042l2.97-2.97H3.75a.75.75 0 0 1 0-1.5h7.44L8.22 4.03a.75.75 0 0 1 0-1.06Z"></path></svg>
</a>
<div class="footer" role="contentinfo" id="steamdb-org" itemscope="" itemtype="http://schema.org/Organization" itemprop="publisher">
<meta itemprop="name" content="SteamDB">
<meta itemprop="url" content="https://steamdb.info/">
<meta itemprop="logo" content="https://steamdb.info/static/logos/512px.png">
<div class="footer-container">
<div class="footer-menu hide-small">
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-info" aria-hidden="true"><path d="M0 8a8 8 0 1 1 16 0A8 8 0 0 1 0 8Zm8-6.5a6.5 6.5 0 1 0 0 13 6.5 6.5 0 0 0 0-13ZM6.5 7.75A.75.75 0 0 1 7.25 7h1a.75.75 0 0 1 .75.75v2.75h.25a.75.75 0 0 1 0 1.5h-2a.75.75 0 0 1 0-1.5h.25v-2h-.25a.75.75 0 0 1-.75-.75ZM8 6a1 1 0 1 1 0-2 1 1 0 0 1 0 2Z"></path></svg> About</li>
<li><a href="/blog/">Blog</a></li>
<li><a href="/discord/">Discord</a></li>
<li><a href="/discord/bot/">Discord Bot</a></li>
<li><a href="/faq/">FAQ &amp; help</a></li>
<li><a href="/donate/">Donate &amp; contribute</a></li>
<li><a href="/extension/">Browser extension</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-search" aria-hidden="true"><path d="M10.68 11.74a6 6 0 0 1-7.922-8.982 6 6 0 0 1 8.982 7.922l3.04 3.04a.749.749 0 0 1-.326 1.275.749.749 0 0 1-.734-.215ZM11.5 7a4.499 4.499 0 1 0-8.997 0A4.499 4.499 0 0 0 11.5 7Z"></path></svg> Tools</li>
<li><a href="/sales/">Sales</a></li>
<li><a href="/charts/">Charts</a></li>
<li><a href="/calculator/">Calculator</a></li>
<li><a href="/patchnotes/">Patch notes</a></li>
<li><a href="/search/">Search</a></li>
<li><a href="/instantsearch/">Instant search</a></li>
<li><a href="/sales/history/">When is next sale?</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-sync" aria-hidden="true"><path d="M1.705 8.005a.75.75 0 0 1 .834.656 5.5 5.5 0 0 0 9.592 2.97l-1.204-1.204a.25.25 0 0 1 .177-.427h3.646a.25.25 0 0 1 .25.25v3.646a.25.25 0 0 1-.427.177l-1.38-1.38A7.002 7.002 0 0 1 1.05 8.84a.75.75 0 0 1 .656-.834ZM8 2.5a5.487 5.487 0 0 0-4.131 1.869l1.204 1.204A.25.25 0 0 1 4.896 6H1.25A.25.25 0 0 1 1 5.75V2.104a.25.25 0 0 1 .427-.177l1.38 1.38A7.002 7.002 0 0 1 14.95 7.16a.75.75 0 0 1-1.49.178A5.5 5.5 0 0 0 8 2.5Z"></path></svg> Tracking</li>
<li><a href="/apps/">Apps</a></li>
<li><a href="/subs/">Packages</a></li>
<li><a href="/bundles/">Bundles</a></li>
<li><a href="/history/">History</a></li>
<li><a href="/pricechanges/">Price changes</a></li>
<li><a href="/upcoming/free/">Free promotions</a></li>
<li><a href="/tech/">Technologies &amp; built with</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-database" aria-hidden="true"><path d="M1 3.5c0-.626.292-1.165.7-1.59.406-.422.956-.767 1.579-1.041C4.525.32 6.195 0 8 0c1.805 0 3.475.32 4.722.869.622.274 1.172.62 1.578 1.04.408.426.7.965.7 1.591v9c0 .626-.292 1.165-.7 1.59-.406.422-.956.767-1.579 1.041C11.476 15.68 9.806 16 8 16c-1.805 0-3.475-.32-4.721-.869-.623-.274-1.173-.62-1.579-1.04-.408-.426-.7-.965-.7-1.591Zm1.5 0c0 .133.058.318.282.551.227.237.591.483 1.101.707C4.898 5.205 6.353 5.5 8 5.5c1.646 0 3.101-.295 4.118-.742.508-.224.873-.471 1.1-.708.224-.232.282-.417.282-.55 0-.133-.058-.318-.282-.551-.227-.237-.591-.483-1.101-.707C11.102 1.795 9.647 1.5 8 1.5c-1.646 0-3.101.295-4.118.742-.508.224-.873.471-1.1.708-.224.232-.282.417-.282.55Zm0 4.5c0 .133.058.318.282.551.227.237.591.483 1.101.707C4.898 9.705 6.353 10 8 10c1.646 0 3.101-.295 4.118-.742.508-.224.873-.471 1.1-.708.224-.232.282-.417.282-.55V5.724c-.241.15-.503.286-.778.407C11.475 6.68 9.805 7 8 7c-1.805 0-3.475-.32-4.721-.869a6.15 6.15 0 0 1-.779-.407Zm0 2.225V12.5c0 .133.058.318.282.55.227.237.592.484 1.1.708 1.016.447 2.471.742 4.118.742 1.647 0 3.102-.295 4.117-.742.51-.224.874-.47 1.101-.707.224-.233.282-.418.282-.551v-2.275c-.241.15-.503.285-.778.406-1.247.549-2.917.869-4.722.869-1.805 0-3.475-.32-4.721-.869a6.327 6.327 0 0 1-.779-.406Z"></path></svg> Database</li>
<li><a href="/tags/">User tags</a></li>
<li><a href="/calendar/">Release calendar</a></li>
<li><a href="/upcoming/">Upcoming releases</a></li>
<li><a href="/upcoming/mostfollowed/">Most followed upcoming</a></li>
<li><a href="/stats/releases/">Game releases by year</a></li>
<li><a href="/badges/">Profile badges</a></li>
<li><a href="/developers/">Developers</a></li>
<li><a href="/publishers/">Publishers</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-number" aria-hidden="true"><path d="M9 4.75A.75.75 0 0 1 9.75 4h4a.75.75 0 0 1 .53 1.28l-1.89 1.892c.312.076.604.18.867.319.742.391 1.244 1.063 1.244 2.005 0 .653-.231 1.208-.629 1.627-.386.408-.894.653-1.408.777-1.01.243-2.225.063-3.124-.527a.751.751 0 0 1 .822-1.254c.534.35 1.32.474 1.951.322.306-.073.53-.201.67-.349.129-.136.218-.32.218-.596 0-.308-.123-.509-.444-.678-.373-.197-.98-.318-1.806-.318a.75.75 0 0 1-.53-1.28l1.72-1.72H9.75A.75.75 0 0 1 9 4.75Zm-3.587 5.763c-.35-.05-.77.113-.983.572a.75.75 0 1 1-1.36-.632c.508-1.094 1.589-1.565 2.558-1.425 1 .145 1.872.945 1.872 2.222 0 1.433-1.088 2.192-1.79 2.681-.308.216-.571.397-.772.573H7a.75.75 0 0 1 0 1.5H3.75a.75.75 0 0 1-.75-.75c0-.69.3-1.211.67-1.61.348-.372.8-.676 1.15-.92.8-.56 1.18-.904 1.18-1.474 0-.473-.267-.69-.587-.737ZM5.604.089A.75.75 0 0 1 6 .75v4.77h.711a.75.75 0 0 1 0 1.5H3.759a.75.75 0 0 1 0-1.5H4.5V2.15l-.334.223a.75.75 0 0 1-.832-1.248l1.5-1a.75.75 0 0 1 .77-.037Z"></path></svg> Rankings</li>
<li><a href="/stats/gameratings/">Top rated games</a></li>
<li><a href="/badge/13/">Top game owners</a></li>
<li><a href="/stats/toplevels/">Top profile levels</a></li>
<li><a href="/stats/globaltopsellers/">Top selling games</a></li>
<li><a href="/topsellers/">Top selling games last week</a></li>
<li><a href="/stats/mostfollowed/">Most followed games</a></li>
<li><a href="/stats/mostwished/">Most wishlisted games</a></li>
</ul>
<ul>
<li class="navigation-heading"><svg width="16" height="16" viewBox="0 0 16 16" class="octicon octicon-person" aria-hidden="true"><path d="M10.561 8.073a6.005 6.005 0 0 1 3.432 5.142.75.75 0 1 1-1.498.07 4.5 4.5 0 0 0-8.99 0 .75.75 0 0 1-1.498-.07 6.004 6.004 0 0 1 3.431-5.142 3.999 3.999 0 1 1 5.123 0ZM10.5 5a2.5 2.5 0 1 0-5 0 2.5 2.5 0 0 0 5 0Z"></path></svg> Account</li>
<li><a href="/watching/1235067451/">Watch list</a></li>
<li><a href="/sales/?displayOnly=OwnedGames&amp;min_reviews=0&amp;min_rating=0&amp;min_discount=0">Your games</a></li>
<li><a href="/freepackages/">Free packages</a></li>
<li><a href="/user/">Settings</a></li>
<li><a href="/signout/">Sign out</a></li>
<li>&nbsp;</li>
<li><a href="https://steamstat.us/">Steam Status</a></li>
<li><a href="https://steamapi.xpaw.me/">Steam Web API</a></li>
</ul>
</div>
<div class="footer-copyright">
<div>
<p>
SteamDB is a hobby project and is not affiliated with Valve or Steam. All times on the site are UTC.
<a href="/faq/#what-steamdb-is-not">Fair use disclaimer</a> Â·
<a href="/faq/#privacy">Privacy</a>
</p>
<p>Steam and the Steam logo are trademarks of Valve Corporation. All other trademarks are property of their respective owners.</p>
</div>
<div class="footer-buttons">
<a class="btn-social tooltipped tooltipped-n" aria-label="Discord" href="/discord/" rel="nofollow">
<svg viewBox="0 0 16 16" width="24" height="24" class="btn-social-icon" aria-hidden="true">
<path d="M13.54 2.9a13.2 13.2 0 00-3.25-1c-.02 0-.04 0-.05.02-.15.25-.3.58-.41.83a12.18 12.18 0 00-3.66 0 8.4 8.4 0 00-.41-.83.05.05 0 00-.05-.02 13.16 13.16 0 00-3.28 1.03 13.5 13.5 0 00-2.34 9.14c1.37 1 2.7 1.61 4 2.02.01 0 .04 0 .05-.02.3-.42.58-.86.82-1.33a.05.05 0 00-.03-.07 8.76 8.76 0 01-1.25-.6.05.05 0 010-.08l.24-.2a.05.05 0 01.05 0 9.46 9.46 0 008.05 0h.05l.25.2c.03.02.02.07 0 .09-.4.23-.82.43-1.26.59a.05.05 0 00-.02.07c.24.47.51.91.81 1.33.02.02.04.03.06.02a13.23 13.23 0 004.02-2.06 13.41 13.41 0 00-2.39-9.12zm-8.2 7.31c-.78 0-1.43-.72-1.43-1.6 0-.9.64-1.62 1.44-1.62.8 0 1.45.73 1.43 1.61 0 .89-.63 1.61-1.43 1.61zm5.32 0c-.79 0-1.43-.72-1.43-1.6 0-.9.63-1.62 1.43-1.62.81 0 1.45.73 1.44 1.61 0 .89-.63 1.61-1.44 1.61z"></path>
</svg>
</a>
<a class="btn-social tooltipped tooltipped-n" aria-label="Twitter" href="https://twitter.com/intent/follow?screen_name=SteamDB" itemprop="sameAs">
<svg viewBox="0 0 24 24" width="24" height="24" class="btn-social-icon" aria-hidden="true">
<path d="M18.901 1.153h3.68l-8.04 9.19L24 22.846h-7.406l-5.8-7.584-6.638 7.584H.474l8.6-9.83L0 1.154h7.594l5.243 6.932ZM17.61 20.644h2.039L6.486 3.24H4.298Z"></path>
</svg>
</a>
<a class="btn-social tooltipped tooltipped-n" aria-label="Mastodon" href="https://infosec.exchange/@SteamDB" itemprop="sameAs" rel="me">
<svg viewBox="0 0 38 38" width="24" height="24" class="btn-social-icon" aria-hidden="true">
<path d="M36.4 8.4c-.6-4-4.1-7.2-8.3-7.8a67 67 0 0 0-9.6-.5c-6.3 0-7.7.4-8.4.5C6.1 1.2 2.3 4 1.4 8c-.4 2-.5 4.2-.4 6.2.1 3 .1 5.8.4 8.7.2 2 .5 3.9 1 5.8.8 3.4 4.3 6.3 7.7 7.5a21.2 21.2 0 0 0 12.6.3c1-.3 2-.6 2.8-1.2v-2.9a32 32 0 0 1-7.4.8c-4.3 0-5.5-2-5.8-2.8a8.6 8.6 0 0 1-.5-2.3h.1c2.4.5 4.8.8 7.3.8h1.7c2.5 0 5-.2 7.5-.7h.2c3.8-.7 7.4-3 7.8-8.7V17c0-.8.3-5.7 0-8.6Zm-5.9 14.2h-4V13c0-2-.8-3-2.6-3-2 0-2.9 1.1-2.9 3.6v5.3h-4v-5.3c0-2.5-1-3.7-2.9-3.7-1.7 0-2.6 1-2.6 3v9.7h-4v-10a7 7 0 0 1 1.6-4.8C10 6.6 11.6 6 13.4 6c2 0 3.6.8 4.6 2.4l1 1.6 1-1.6C21 6.8 22.6 6 24.7 6c1.7 0 3.2.6 4.2 1.8a7 7 0 0 1 1.6 4.9v10Z"></path>
</svg>
</a>
<a class="btn-social tooltipped tooltipped-n" aria-label="GitHub" href="https://github.com/SteamDatabase" itemprop="sameAs">
<svg viewBox="0 0 16 16" width="24" height="24" class="btn-social-icon" aria-hidden="true">
<path fill-rule="evenodd" d="M8 0C3.58 0 0 3.58 0 8c0 3.54 2.29 6.53 5.47 7.59.4.07.55-.17.55-.38 0-.19-.01-.82-.01-1.49-2.01.37-2.53-.49-2.69-.94-.09-.23-.48-.94-.82-1.13-.28-.15-.68-.52-.01-.53.63-.01 1.08.58 1.23.82.72 1.21 1.87.87 2.33.66.07-.52.28-.87.51-1.07-1.78-.2-3.64-.89-3.64-3.95 0-.87.31-1.59.82-2.15-.08-.2-.36-1.02.08-2.12 0 0 .67-.21 2.2.82.64-.18 1.32-.27 2-.27.68 0 1.36.09 2 .27 1.53-1.04 2.2-.82 2.2-.82.44 1.1.16 1.92.08 2.12.51.56.82 1.27.82 2.15 0 3.07-1.87 3.75-3.65 3.95.29.25.54.73.54 1.48 0 1.07-.01 1.93-.01 2.2 0 .21.15.46.55.38A8.013 8.013 0 0 0 16 8c0-4.42-3.58-8-8-8z"></path>
</svg>
</a>
<a class="btn-social tooltipped tooltipped-n" aria-label="Steam Group" href="https://steamcommunity.com/groups/steamdb" itemprop="sameAs">
<svg viewBox="0 0 16 16" width="24" height="24" class="btn-social-icon" aria-hidden="true">
<path d="M8 0a8 8 0 00-8 7.47c.07.1.13.21.18.32l4.15 1.67a2.2 2.2 0 011.31-.36l1.97-2.8v-.04c0-1.65 1.37-3 3.05-3a3.03 3.03 0 013.05 3 3.03 3.03 0 01-3.12 3l-2.81 1.97c0 .3-.05.6-.17.9a2.25 2.25 0 01-4.23-.37L.4 10.56A8.01 8.01 0 108 0zm2.66 4.27c-1.12 0-2.03.9-2.03 2s.91 1.99 2.03 1.99c1.12 0 2.03-.9 2.03-2s-.9-2-2.03-2zm0 .5c.85 0 1.53.66 1.53 1.49s-.68 1.5-1.53 1.5c-.84 0-1.52-.67-1.52-1.5s.68-1.5 1.52-1.5zM5.57 9.6c-.22 0-.43.04-.62.11l1.02.42c.65.26.95.99.68 1.62-.27.63-1 .93-1.65.67l-1-.4a1.73 1.73 0 003.13-.08c.18-.42.18-.88.01-1.3A1.69 1.69 0 005.57 9.6z"></path>
</svg>
</a>
</div>
</div>
</div>
</div>
</div>
<div id="js-notifications">
<noscript>
<div class="toast-header"><b>Enable JavaScript</b></div>
<div class="toast-body">For SteamDB to function correctly, JavaScript should be enabled.</div>
</noscript>
</div>
<div><div id="js-hover" hidden=""></div></div>
<div id="js-screenshots">
<div class="screenshot-header">
<span>
<a href="#" class="screenshot-name"></a>
</span>
<span>
<button type="button" class="btn btn-sm btn-primary screenshot-prev">â† Prev</button>
<span class="screenshot-index"></span>
<button type="button" class="btn btn-sm btn-primary screenshot-next">Next â†’</button>
</span>
<span>
<a href="#" target="_blank" class="btn btn-sm screenshot-open hide-small">Open link</a>
<button type="button" class="btn btn-sm btn-danger screenshot-close">Close<span class="hide-small"> (Esc)</span></button>
</span>
</div>
<div class="screenshot-contain"></div>
</div>
<script defer="" src="https://static.cloudflareinsights.com/beacon.min.js/v84a3a4012de94ce1a686ba8c167c359c1696973893317" integrity="sha512-euoFGowhlaLqXsPWQ48qSkBSCFs3DPRyiwVu3FjR96cMPx+Fr+gpWRhIafcHwqwCqWS42RZhIudOvEI+Ckf6MA==" nonce="" data-cf-beacon="{&quot;rayId&quot;:&quot;84d5b11ffdd219f6&quot;,&quot;version&quot;:&quot;2024.1.0&quot;,&quot;token&quot;:&quot;91efbe05a59742ddadf5d555608bbe98&quot;}" crossorigin="anonymous"></script>
</body>
'''
```

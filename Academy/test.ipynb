```python
# 모듈 세팅
import pygame
import sys

from pygame import Rect

from enum import Enum

class TILE_INFO(Enum):
    EMPTY = 0
    WALL = 1
    
class TILE:
    def __init__(self, label : tuple, type : TILE_INFO):
        self.index_x , self.index_y = (0,0)
        self.index_x , self.index_y = label
        
        self.type = TILE_INFO.EMPTY

# 매크로
SCREEN_SIZE_X = 1280
SCREEN_SIZE_Y = 768

STATIC_FRAME = 30

TILE_SIZE = 32

# 윈도우 전역 
surface = None
debug = True

scene = 0

# 메인

def tileInit(tile_x_count, tile_y_count):    
    return [TILE(label=(x,y)) for y in range(tile_y_count) for x in range(tile_x_count)]

def mainGame(surface):
    mouse_x,mouse_y = pygame.mouse.get_pos()
    
    mouse_check_x = int(mouse_x/TILE_SIZE)
    mouse_check_y = int(mouse_y/TILE_SIZE)

    pygame.draw.rect(surface, (255,255,0),(mouse_check_x * TILE_SIZE, mouse_check_y * TILE_SIZE, TILE_SIZE, TILE_SIZE))
    
    debugfont = pygame.font.SysFont(None,36) # 텍스트 세팅
    text = debugfont.render('Mouse X : {} , Mouse Y : {}'.format(mouse_check_x, mouse_check_y),True, (0,0,255))
    surface.blit(text,(800,100))

def debugInit():
    # 디버그용 글로벌 변수
    global tile_x_count, tile_y_count,tile_number_list
    global small_font
    tile_x_count = int(SCREEN_SIZE_X/TILE_SIZE)
    tile_y_count = int(SCREEN_SIZE_Y/TILE_SIZE)
    
    tile_number_list = [(x , y) for x in range(tile_x_count) for y in range(tile_y_count)]
    
    small_font = pygame.font.SysFont(None,16)
    

def debugGame(surface):
    global scene
    # 디버그용 글로벌 변수
    global tile_x_count, tile_y_count,tile_number_list
    global small_font
    # X 타일
    for x in range(tile_x_count):
        pygame.draw.line(surface,(64,64,64),(x * TILE_SIZE, 0),(x *TILE_SIZE, SCREEN_SIZE_Y))
    # Y 타일
    for y in range(tile_y_count):
        pygame.draw.line(surface,(64,64,64),(0, y * TILE_SIZE), (SCREEN_SIZE_X, y * TILE_SIZE))    
    
    for y in range(tile_y_count):
        for x in range(tile_x_count):
            num_image = small_font.render('{}/{}'.format(x,y), True,(0,0,255))
            surface.blit(num_image, (x * TILE_SIZE + 6, y * TILE_SIZE + 6))
    
    # 텍스트 기본 세팅    
    debugfont = pygame.font.SysFont(None,36) # 텍스트 세팅
    text = debugfont.render('Debug Mode {}'.format(scene),True, (0,0,255))
    surface.blit(text,(1000,0))
    
    

#####################################################
## 게임 기본 세팅

def initGame():
    global surface
    
    pygame.init()
    surface = pygame.display.set_mode((SCREEN_SIZE_X, SCREEN_SIZE_Y))    
    pygame.display.set_caption('국제 AI 1기 학습 메타버스')

def runGame():
    global surface
    global scene, debug
    
    onGame = False
    
    # pygame 타이머
    clock = pygame.time.Clock()

    mainInit()
    debugInit()

    while not onGame:
        event = pygame.event.poll()
        
        if event.type in [pygame.QUIT]:
            onGame = True
        
        surface.fill((255,255,255))
        
        # 키 입력
        if event.type in [pygame.KEYDOWN]:
            if event.key == pygame.K_ESCAPE:
                onGame = True    

            if event.key == pygame.K_5:
                if debug == True:
                    debug = False
                elif debug == False:
                    debug = True
            
            if debug == True:
                if event.key == pygame.K_9:
                    if scene > 0:
                        scene -= 1
                if event.key == pygame.K_0:        
                    if scene < 10:
                        scene += 1
            
        mainGame(surface)
        
        if debug == True:
            debugGame(surface)
        
        
        pygame.display.update()
        clock.tick(STATIC_FRAME)
    
    pygame.quit()
    sys.exit()
    

# 메인
if __name__ == '__main__':
    initGame()
    runGame()
```
